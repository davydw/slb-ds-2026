<details>
  <summary>What is the formula of a <b>log-odd (logit)</b> for a probability p?</summary>

The formula is:
$$ logit(p) = log(\frac{p}{1-p}) $$

In other words, it is the **log of the odds ratio** (success/failure ratio).

It is also called the **logit function**
</details>

<details>
  <summary>How would a 1 dollar fare increase impact the log odds of surviving?

```python
model3 = smf.logit(formula='survived ~ fare', data=titanic).fit()
model3.params
# Intercept   -0.941330
# fare         0.015197
```
</summary>

Increasing the fare by 1 dollar increases the log odds of surviving by 0.015.

Higher log odds also means higher odds, and thus the odds of surviving increase.

Higher odds of surviving also means higher probability of surviving.

So, a 1 dollar fare increase would increase the probability of surviving.

In general, an increase in log odds means an increase in probability.
</details>

<details>
  <summary>In which way is the logit transformation interpretable?</summary>

The logit transformation is interpretable in terms of **odd_ratios** because:

$$log(odd_{1}) - log(odd_{2}) = log(\frac{odd_{1}}{odd_{2}}) $$
</details>

```python
model2 = smf.logit(formula='survived ~ C(pclass)', data=titanic).fit()
model2.params
# Intercept         0.530628
# C(pclass)[T.2]   -0.639431
# C(pclass)[T.3]   -1.670399
```  
  
  </summary>

* 0.53 is the log odds of surviving for passengers that were in first class
* -0.63 is the **decrease** in log odds of survival for 2nd class passenger, **relative** to passenger of first class
</details>

<details>
  <summary>What are the <b>signs of multicollinearity</b> when using OLS?</summary>

Multicollinearity occurs when your model includes multiple factors that are correlated not just to your target, but also to each other.

Signs of multicollinearity include **large standard errors** combined with **high correlation** between independent variables and high correlation between estimated coefficients.
</details>

<details>
  <summary>Given a probability of success of 0.8, what are the odds of success?</summary>

First, let’s compute the odds ratio: 0.8 / 0.2 = 4.

We can now say that the odds of success are 4 to 1.
</details>

<details>
  <summary>How many solutions to OLS are there if two features are collinear?</summary>

* There is **no single solution** to OLS if two features are collinear.
* There can also be an infinite number of solutions for a non-full-rank matrix of features
</details>

<details>
  <summary>When using linear regression, how can you check for <b>collinearity</b>?</summary>

You should look at the **correlation matrix** and compute the **rank of the feature matrix**. The correlation matrix will show collinearity between 2 features. If the rank of the feature matrix is less than the number of features, it means there is multicollinearity.

This will only detect strict multicollinearity. So you should also perform a VIF (Variance Inflation Factor) test.
</details>

<details>
  <summary>For which <b>rank value</b> of the matrix of features X is X <b>invertible</b>?</summary>


 $X^T X$ is **invertible** if and only if $rank(X)$ is  **equal to the number of features**.
</details>

<details>
  <summary>Suppose a binary outcome variable <i>Y</i> with probability <i>p</i> and features X<sub>1</sub>, X<sub>2</sub>, ..., X<sub>n</sub>. What is the <b>formula of the logistic regression</b>?</summary>

$$ log(\frac{p}{1-p}) = \beta_{0} + \beta_{1} X_1 + ... + \beta_{n} X_n$$

using the Maximum Likelihood Estimated method to estimate $\beta$
</details>

<details>
  <summary>What is a Logistic Regression used for?</summary>

A logistic regression is used to **predict a binary outcome**.
</details>

<details>
  <summary>Is the R<sup>2</sup> a good measure for logistic regression?</summary>

In a logistic regression, we can’t calculate R<sup>2</sup>, but we can calculate the pseudo-R<sup>2</sup>.

It is **not as descriptive** as the R<sup>2</sup> from linear regression:

* It is **valid** in comparing multiple models trying to predict the same outcome with the same dataset
* It is **invalid** in comparing different datasets
</details>

<details>
  <summary>The logit transformation maps probability range [0-1] into which interval?</summary>

$$[-\infty, +\infty]$$
</details>
