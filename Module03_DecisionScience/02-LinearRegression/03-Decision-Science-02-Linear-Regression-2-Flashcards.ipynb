# 03-DecisionScience_02-Linear-Regression-2-Flashcards

<details>
  <summary>What does a F-statistic of close to <i>$1</i> mean?</summary>

It means that you **cannot rule out** that the linear regression is **not statistically significant**.
</details>

<details>
  <summary>When using linear regression, how can you check **inference conditions**?</summary>

You should **plot residuals** to make sure they are randomly distributed, centered on 0, with a constant variance and that you cannot spot some correlation.
</details>

<details>
  <summary>Does a high R<sup>2</sup> imply statistical significance?</summary>

No, you have to compute the **p-value** to show statistical significance.
</details>

<details>
  <summary>In the scope of a <b>multivariate regression</b>, why would you look into partial regression plots?

</summary>

To visualize the effect of a particular explaining variable on the dependent variable while holding all other explaining variables constant.
</details>

<details>
  <summary>When using linear regression, how can you check <b>goodness-of-fit</b>?</summary>

You should compute the $R^2$ score to make sure that the model explains a good deal of the observed variance of the dependent variable.
</details>

<details>
  <summary>After training an OLS statsmodels, you get the following params (<i>y</i> being the weight of the cars in kgs):

```python
 Intercept     984.500327
 horsepower     19.078162
 ```
 How would you interpret those results?  

  </summary>

* For each increase of 1 `horsepower`, car `weights` increase on average by 19 kg
* A car with 0 `horsepower` should weigh 984kg
</details>

<details>
  <summary>What’s the range of values taken by the R<sup>2</sup></summary>

The $R^2$ usually ranges from 0 (explains nothing) to 1 (perfect relationship).

Worst case $R^2$ can even be negative if the model is worse than a simple average.
</details>

<details>
  <summary>Given the following linear regression formula, what is the size of the residuals vector u?

 $$ \hat{Y} = \beta_{0} + \beta_{1}X_1 + ... + \beta_{k}X_k + \vec{u} $$
  
  </summary>

The residual vector $\vec{u}$ is of size 
, being the **number of observations**. You can think of it as the difference between the actual value for each observation and the value given by the linear regression.

The residual vector is **not** of size *k*, which is the number of **features**.
</details>

<details>
  <summary>What is <b>Ordinary Least Square</b>?</summary>

In statistics, ordinary least squares (OLS) is a type of **linear regression model** minimizing the sum of the squares of the differences between the observed dependent variable in the given dataset and those predicted by the linear function.
</details>

<details>
  <summary>What is the R-squared(R<sup>2</sup>)?</summary>

The **coefficient of determination** or **r-squared** is the proportion of the variance in the dependent variable that is predictable from variance of the features.
</details>

<details>
  <summary>An Ordinary Least Square regression minimizes which distance?</summary>

The Ordinary Least Square regression minimizes the “natural” $L_2$ **euclidian distance** between the observed datapoints and the regression “line” (a “plane” in 3D).
</details>

<details>
  <summary>When plotting errors reveals a pattern, what does that tell you?</summary>

If a pattern is seen, a factor might be missing in your model!

It is a frequent issue in Time Series (ex: inflation, weekly patterns etc…)
</details>

<details>
  <summary>After training a multivariate OLS statsmodels, you get the following params (<i>y</i>being the weights of the cars in kgs):
  
  ```python
Intercept     528.876711
horsepower      8.231070
cylinders     290.356425

  ```
  </summary>

* Each increase in `horsepower` increases the `weight` by about 8 kgs, holding `cylinders` constant.
* Each increase in `cylinders` increases the `weight` by about 290 kgs, holding `horsepower` constant.
* A car without cylinders or horsepower has a starting weight of about 529 kgs
</details>

<details>
  <summary>How can you visually get an intuition whether it makes sense to use a univariate linear regression?</summary>

Plot a scatterplot of your target $y$ vs. your feature  $X$ and look for a tendency.
</details>

<details>
  <summary>Performing a regression of car <i>weights</i> vs. car <i>horsepower</i>, you get R<sup>2</sup> = 0.74. How do you interpret this value?</summary>

74% of the variance of `weights` is explained by the variance of `horsepower`
</details>

<details>
  <summary>What is the <b>F-statistic</b>?</summary>

The F-statistic or F-test of **overall statistical significance of the regression** indicates whether your linear regression model provides a better fit for the data than a model that contains no independent variables.

It consists of computing the **p-value** assuming the hypothesis that all coefficients are null.
</details>

<details>
  <summary>When using linear regression, how can you check <b>statistical significance</b>?</summary>

You should compute **p-values** and the **F-statistic** to make sure your model generalizes well.
</details>

<details>
  <summary>What does a high value <b>F-statistic</b> mean?</summary>

It means that the regression is **statistically significant**. The higher the better!
</details>
