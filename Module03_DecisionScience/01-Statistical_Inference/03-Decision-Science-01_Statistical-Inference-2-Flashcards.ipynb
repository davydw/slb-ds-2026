<details>
  <summary>When is n large enough for the Central Limit Theorem to converge toward a normal distribution?</summary>

* If n>30
* If n>10  and observations are “non-skewed” and without outliers
* If global population is known to be normally distributed
</details>

<details>
  <summary>What methodology would you use to test an hypothesis?</summary>

1. Create **Null Hypothesis** H<sub>0</sub>
2. Create **Alternative Hypothesis** H<sub>a</sub>
3. Choose a **Significance level** $\alpha$
4. Suppose H<sub>0</sub> is true and **compute the probability of observing a sample mean**
5. Compute the **p-value** and compare it to the significance level
    * If p-value < $\alpha$, then we **reject** the null hypothesis H<sub>0</sub> in favour of the alternative hypothesis H<sub>a</sub>.
    * If p-value > $\alpha$, then we **fail to reject the null hypothesis** H<sub>0</sub>. This doesn’t mean we accept H<sub>0</sub> .
</details>

<details>
  <summary>What is the Central Limit Theorem?</summary>

When independent random variables $X_{1}$, ..., $X_{n}$ with common probability distribution (of mean $\mu$ and std $\sigma$) are added:

Their mean converges towards a normal distribution as the number of sample increases
* centered on the common mean $\mu$
* of standard deviation $\frac{\sigma}{\sqrt{n}}$
 

This holds true whatever the form of the common distribution
</details>

<details>
  <summary>How would you create an “A/B Test” for a new app feature (an experiment design)?</summary>

1. Develop the corresponding feature
2. Create two groups (control group vs. treatment group)
3. Randomly assign users to the treatment group
4. Deploy the feature only to treatment group
5. Collect behavior statistics (e.g. sample mean)
</details>

<details>
  <summary>What is a type I error?</summary>

A type I error is the rejection of a true null hypothesis:
* $H_0$ is actually true
* But $H_0$ is rejected
It is also known as a “false positive”.
</details>

<details>
  <summary>What is a type II error?</summary>

A type II error is the non-rejection of a true null hypothesis:
* $H_0$ is actually false
* But $H_0$ is not rejected
It is also known as a “false negative.
</details>

<details>
  <summary>What is a confidence interval?</summary>

A confidence interval is a range of values, derived from sample data, that is likely to contain the true population parameter.

It provides an estimate of uncertainty, expressed with a specific confidence level (e.g., 95%).
</details>

<details>
  <summary>What is the p-value?</summary>

The p-value or probability value is the probability of obtaining test results at least as extreme as the results actually observed during the test, assuming that the null hypothesis is correct.

$\mathbb{P}(\bar{X} \geq x | H_0)$
</details>

<details>
  <summary>What is statistical inference?</summary>

Statistical inference is the process of using **data analysis** to deduce properties of an **underlying distribution of probability**. Inferential statistical analysis infers **properties of a population**, for example by **testing hypotheses** and **deriving estimates**. It is assumed that the observed dataset is sampled from a larger population.
</details>

<details>
  <summary>Which distribution can you use if your sample size <i>n</i>is not large enough, regarding the Central Limit Theorem?</summary>

You can use a student t-distribution.

Be aware that you still need independent and random sampling.
</details>

<details>
  <summary>You conduct an experiment where you try to measure a statistic <i>mu</i> via random sampling, and you measure a result <i>X_bar</i> from one sample. If you had <b>no prior beliefs</b> for the value of <i>mu</i>, what would be your most plausible estimate for <i>mu</i>? What would you call such an estimate?</summary>

The most plausible estimate for $\mu$ would be the **Maximum Likelihood Estimate** $\bar{X}$
</details>

<details>
  <summary>What is a standard score or z-score?</summary>

The standard or z-score score is the number of standard deviations by which the value of an observation is above or below the mean value of what is expected from its sampling distribution.
</details>

<details>
  <summary>You conduct an experiment where you try to measure a statistic <i>mu</u>via random sampling, and you measure a result <i>X_bar</i> from one sample. If you had a <b>prior belief</b>, modelled by a prior probability density, which formula would you use to update your belief to a <b>posterior probability density?</b> What would you call your new best guess?</summary>


You would use **Bayes Theorem** to update your belief to a *posterior probability density*.
You would call your new best guess the **Maximum A Posteriori Estimate**.
</details>

<details>
  <summary>With a sample size of n, how do you compute the z-score for a sample mean <i>X_bar</i>, assuming the population has a mean <i>mu</i> and a standard deviation <i>sigma</i></summary>

$$Z = \frac{\bar{X}-\mu}{\frac{\sigma}{\sqrt{n}}}$$
</details>

<details>
  <summary>When is n small enough for each draw to be considered independent, even without replacement?</summary>

When the size of the sample represents less than 10% of the global population size
</details>

<details>
  <summary>Prior to tossing a coin, we may have an opinion about the coin fairness. What is this initial belief called?</summary>

* This initial belief is called the **prior probability** $\mathbb{P}(H)$ (the probability of flipping a head)
* If we have no opinion, we model the $\mathbb{P}(H)$ as an uniform distribution over [0,1]
</details>

<details>
  <summary>What is the power of a statistical test?</summary>

Power is the probability that we will **correctly reject the null hypothesis** (if it was correct to reject it)

For example:

* Probability of not missing out on a great feature in A/B testing
* Probability of not missing out on an effective new drug in a clinical trial
* Probability of not making a type II error
</details>

<details>
  <summary>How does the t-distribution shape compare to the normal distribution N(0,1)?
?</summary>

The t-distribution has fatter tails than the normal distribution, meaning that it is more prone to producing values that fall far from its mean.
</details>
