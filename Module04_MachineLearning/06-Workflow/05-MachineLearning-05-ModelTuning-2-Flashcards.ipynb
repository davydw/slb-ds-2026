# 05-MachineLearning-06-Workflow-2-Flashcards

<details>
  <summary>While preprocessing your data, you notice that one of your features has over 80% NaN values. Should you systematically drop the entire feature?</summary>

NO! NaNs can carry information that should not be discarded. Investigate the origin of the NaN first. You can then drop the feature if it turns out to not carry any info.
</details>

<details>
  <summary>Are elements of a pipeline activated sequentially or in parallel?</summary>

Elements of a pipeline are activated sequentially, usually from data preprocessing to modelling.
</details>

<details>
  <summary>Name a few potential causes for overfitting, as well as a few approaches to combat it.</summary>

Overfitting happens when the model learns the training data too closely, almost by heart. When that happens, it is unable to generalize well unseen data.

**Causes**: The model could be overly complex with too many parameters, there could too few training examples, or you are training for too many epochs without proper regularization (SGD and Deep Learning).

**Approaches**: You could add more data for training, use regularization (Ridge or Lasso), tune the model parameters, or try a simpler model.
</details>

<details>
  <summary>What is the difference between a grid search and a random search?

</summary>

* In a grid search, the data scientist specifies which hyperparameter value to test, and each possible combination is tested.
* In a random search, a defined number of random samples are made out of a hyperparameter space and tested for performance. 50 sample iterations has proven to be a good tradeoff between time and efficiency.
</details>

<details>
  <summary>If you perform a 10-fold Cross Validation when training a classification model, it will return 10 accuracy scores. Which one of the 10 scores should you be looking at as indicator of performance?</summary>

You should consider ALL 10 accuracy scores and look at their mean value as performance indicator.
</details>

<details>
  <summary>You are doing a binary classification task, testing a patient to have a virus (e.g. COVID-19), you don’t want to miss any COVID-19 case. It is ok to have some false alarms. Which performance metric should you look at?</summary>

You should be looking at the recall score. The ability to detect occurences of a class of interest.
</details>

<details>
  <summary>Your pipeline is made up of a scaler and a linear model. If the pipeline’s predict method is called, the scaler would be fitted to the data before transforming it. True or False?</summary>

False! On execution of pipeline’s predict method, only the transformers’ `transform` method will be called, using the variables learnt during the original `fit`.
</details>

<details>
  <summary>When preprocessing your data, you should fit two separate standardizers for the training set and test set. True or False?</summary>

False! Standardization performs transformations based on constants of the set it is fit on (mean and standard deviation of the training set). It should be fitted once to the training set and used to transform both the train set and the test set.
</details>

<details>
  <summary>Other than by name, how can you select the features to be passed in a transformer?</summary>

You could do that by type. `.dtypes` can be used to check the types of your dataframe columns. They can be selected in a transformer via the pluggin tool `make_column_selector()`, to which you pass the type you target.
</details>

<details>
  <summary>What are you doing if you evaluate your model on the same data you trained it with?</summary>

You are overfitting! A model’s performance should be evaluated on data points it has not seen during training.
</details>

<details>
  <summary>Name a few potential causes for underfitting, as well as a few approaches to combat it.</summary>

Underfitting occurs when the model is unable to capture and learn a structure in the data.
* **Causes**: The model could be too simplistic, or the data not sufficiently informative.
* **Approaches**: In that case, you could choose a more complex model or do some feature engineering.
</details>

<details>
  <summary>What does the Sklearn tool <b>ColumnTransformer</b> do?</summary>

It allows you to perform parallel preprocessing operations on specific columns and package it as a single transformer. For example, scaling a numerical feature while encoding a categorical one.
</details>
