# [B_Wrapup] Time Series

📈 Time Series can be decomposed, additively or multiplicatively, in three components:
* trend
* seasonality
* residuals - hopefully stable around 0 (➕ )  or 1 (✖️)

✂️  You have to train_test_split manually because of… the temporal nature of a Time Series !

⚠️  You should work on stationary TS. You have multiple ways to achieve it and you can also combine them together:
* de-seasonalizing
* differentiate
* apply a transformation such as the logarithm

👉 You can run the `adfuller test` on a TS. If the outcome is lower than 0.05 :good-style:  it is stationary.

🤓  An `ARMA(p,q)` model is the sum of two Linear Regressions:
* one on `p autoregressive features` (predicting $Y_t$ based on $Y_{t-1}, Y_{t-2}, …, Y_{t - p}$)
* one based on `q errors` of the predictions computed through moving averages (predicting$Y_t$ based on $epsilon_{t-1}, epsilon_{t-2}, …, epsilon_{t - q}$)

🤓  An `ARIMA(p,d,q)` model is an ARMA model which takes into account the fact that you differentiated your TS d times. Tip: use `.diff()`

👀  How do we choose p and q ? 
* PACF gives you p
* ACF gives you q

🚀 You can GridSearch an ARIMA using `auto_arima`
But do not trust it blindly, if you have similar AICs, go for the simplest model !

✅  After fitting an ARIMA, you can plot the predictions using `plot_predict()` and forecast using `forecast()`

⚠️  In the anti-diabetic drug sales, we:
1. removed the seasonality (dividing the TS by `seasonal_mul`
2. applied the logarithm function to linearize the TS
When you plot your forecasts, do not forget to apply the inverse operations !
1. exponentiate your TS, your forecasts, your lower and upper bounds
2. re-introduce the seasonality with a multiplication
