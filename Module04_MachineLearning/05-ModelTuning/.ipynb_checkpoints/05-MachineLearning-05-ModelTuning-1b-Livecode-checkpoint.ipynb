# Hyperparameter tuning in a nutshell

## 📚Libraries

```python
### DATA MANIPULATION
import pandas as pd
pd.options.display.max_columns = None
### DATAVIZ
import matplotlib.pyplot as plt
import seaborn as sns
### DATASET
from sklearn.datasets import load_diabetes
### HOLD METHOD
from sklearn.model_selection import train_test_split
### MODEL
from sklearn.linear_model import ElasticNet
### HYPERPARAMETER TUNING
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
### STATISTICAL TOOLS
from scipy import stats
```

```python
X, y = load_diabetes(return_X_y = True, as_frame = True)
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## 👩🏾‍🔬 GridSearchCV

```python
param_grid = {
    'alpha' : [0.01, 0.1, 1],
    'l1_ratio' : [0.2, 0.5, 0.6, 0.8]
}    
```

```python
grid_search = GridSearchCV(estimator = ElasticNet(),
                           param_grid = param_grid,
                           cv = 5,
                           scoring = ['r2', 'neg_mean_squared_error'],
                           refit = 'r2',
                           n_jobs = -1)

grid_search
```

```python
%%time
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_score_
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
# grid_search.predict(X_test)
```

```python
pd.DataFrame(grid_search.cv_results_).sort_values(by="rank_test_r2").head()
```

```python
pd.DataFrame(grid_search.cv_results_).sort_values(by="mean_test_neg_mean_squared_error").head()
```

### 🧑🏻‍💻 RandomizedSearchCV vs. GridSearchCV

```python
param_distributions = {
    'alpha': [0.001, 0.01, 0.1, 1],
    'l1_ratio' : stats.uniform(0,1)
    
    
}
```

```python
random_search = RandomizedSearchCV(estimator = ElasticNet(),
                                   param_distributions = param_distributions,
                                   scoring = ['r2','neg_mean_squared_error'],
                                   refit = 'r2',
                                   cv = 5,
                                   n_jobs = -1,
                                   n_iter =10)

random_search
```

```python
%%time
random_search.fit(X_train, y_train)
```

```python
random_search.best_score_
```

```python
random_search.best_params_
```

```python
random_search.best_estimator_
```

```python
pd.DataFrame(random_search.cv_results_).sort_values(by="rank_test_r2")
```
