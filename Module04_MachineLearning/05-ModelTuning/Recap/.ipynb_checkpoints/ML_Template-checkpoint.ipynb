### (2.1) 🔢 Numerical values: Scaling

```python
X_train.dtypes.value_counts()
```

```python
# X_train_num = X_train.select_dtypes(include = ["int64", "float64"])
X_train_num = X_train.select_dtypes(include = np.number)
X_train_num.head()
```

```python
len(X_train_num.columns)
```

```python
for variable in list(X_train_num.columns):
    print("-"*50)

    fig, ax = plt.subplots(1,2,figsize=(15,5))

    ax[0].set_title(f"Distribution of the {variable}")
    sns.histplot(data = X_train_num, x = f"{variable}", kde=True, ax = ax[0])

    ax[1].set_title(f"Boxplot of the {variable}")
    sns.boxplot(data = X_train_num, x = f"{variable}", ax=ax[1])

    plt.show();
```

```python
robust_features = ["Flight Distance"]
standard_features = ["Age", ]
minmax_features = ["Seat comfort", "Departure/Arrival time convenient", "Food and drink", 
                   "Gate location", "Inflight wifi service", "Inflight entertainment",
                   "Online support", "Ease of Online booking", "On-board service",
                   "Leg room service", "Baggage handling", "Checkin service",
                   "Cleanliness", "Online boarding", "Departure Delay in Minutes",
                   "Arrival Delay in Minutes"]
```

```python
scalers = ColumnTransformer([
    ("robust", RobustScaler(), robust_features),
    ("standard", StandardScaler(), standard_features),
    ("mm", MinMaxScaler(), minmax_features),    
])

scalers
```

```python
num_preproc = Pipeline([
    ("num_imputer", SimpleImputer(strategy="median")),
    ("scalers", scalers)
]).set_output(transform="pandas")

num_preproc
```

```python
X_train_num.head()
```

```python
num_preproc.fit(X_train_num)

X_train_num_scaled = num_preproc.transform(X_train_num)
X_test_num_scaled = num_preproc.transform(X_test_num)
```
