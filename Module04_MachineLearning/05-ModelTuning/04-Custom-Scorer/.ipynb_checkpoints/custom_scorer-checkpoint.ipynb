# Custom scorer

````python
import pandas as pd
````

Let's create a dummy "wine" dataset of 1000 wines, 10 features, and 3 classes (0=bad, 1=medium, 2=good wine)

````python
from sklearn.datasets import make_classification

X, y = make_classification(
    n_samples=1000,
    n_features=10,
    n_classes=3,
    n_clusters_per_class=1,
    weights=[0.6, 0.3, 0.1],
    random_state=0
)
````

````python
pd.DataFrame(X)
````

````python
pd.Series(y).value_counts()
````

-----
❓ Our objective is to train a model which **maximizes prediction precision for the good wines (y=2) only**.  

We don't want any customers to be dissatisfied!

----

## Optimizing for accuracy 

````python
# Split train/test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 0)
````

❓ Randomized Search CV the best SVC model measured on the `accuracy`
(you can limit yourself to `rbf` kernels to start with)

````python
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV
from scipy import stats
````

````python
%%time

# Instantiate model
model = SVC()

# Hyperparameter search space
search_space = {
    'kernel': ['poly'],
    'C': stats.uniform(0.01, 1000),
    'degree': stats.randint(1,10),
}

# Instantiate Randomized Search
search = RandomizedSearchCV(
    model,
    search_space,
    n_jobs=-1,
    scoring='accuracy',
    cv=5, n_iter=100,
    verbose=1,
    random_state = 0)

# Fit to search for the best hyperparameters
search = search.fit(X_train, y_train)

# Display the best model
search.best_estimator_
````

❓ Print classification report on the test set

````python
from sklearn.metrics import classification_report
print(classification_report(y_test, search.best_estimator_.predict(X_test)))
````

☝️ Not good enough, we want to focus **only** on class 2

## Custom scoring function

Let's make our own custom metric which returns the precision of class "2" 

reminder: ```precision2 = TP2/(TP2+FP2)``` 

````python
def my_custom_metric(y_true, y_pred):
    # $CHALLENGIFY_BEGIN
    # Number of predictions for the chosen label (true positives + false positives)
    TP_and_FP = sum([value == 2 for value in y_pred])

    # Number of true positives
    TP = sum([(t == 2 and p == 2) for t, p in zip(y_true, y_pred)])

    # Calculate precision
    precision = (TP / TP_and_FP) if TP_and_FP != 0 else 0

    return precision
    # $CHALLENGIFY_END

_true = [0,0,1,2,2,1]
_pred = [0,1,1,2,1,0]

my_custom_metric(_true, _pred)  # should return 1
````

````python
# Bonus - one liner leveraging sklearn precision_score
from sklearn.metrics import precision_score
def my_custom_metric_2(y_true, y_pred):
    return precision_score(y_true, y_pred, average=None, zero_division=0)[2].tolist()

_true = [0,0,1,2,2,1]
_pred = [0,1,1,2,1,0]

my_custom_metric_2(_true, _pred)
````

## Optimizing hyper-params for our custom scoring function

````python
# Can you use this function in  sklearn (will crash)
# RandomizedSearchCV(SVC(),
#                    search_space,
#                    n_iter = 100,
#                    cv=5,
#                    scoring=my_custom_metric,
#                    random_state = 0)\
# .fit(X_train, y_train)
````

### `Make_scorer`

We need to transform our "metric" into a "sklearn scorer method"

````python
from sklearn.metrics import make_scorer

my_custom_scorer = make_scorer(my_custom_metric)
````

### RandomizedSearchCV with custom scorer

````python
search = RandomizedSearchCV(
    SVC(),
    search_space,
    n_iter = 100,
    cv=5,
    scoring=my_custom_scorer,
    random_state = 0)

search.fit(X_train, y_train)

search.best_estimator_
````

❓ Print classification report on the test set and compare with the previous one

````python
from sklearn.metrics import classification_report

print(classification_report(y_test, search.best_estimator_.predict(X_test)))
````

✅  We improved our precision for class 2 from 0.69 up to 0.73, but at the detriment of overall accuracy!
