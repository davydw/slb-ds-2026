# 05-MachineLearning-05-ModelTuning-2-Flashcards

<details>
  <summary>What does the parameter C of an SVM regulate?</summary>

Parameter `C` regulates the trade-off between accurate classifiation and generalisation. It influences the separating hyperplane by allowing it more or less misclassification.
</details>

<details>
  <summary>Imagine you are training a Linear Regression model. You plot its Learning Curves and they are yet to converge: you are overfitting. You do not have any more data at your disposal. What do you do?</summary>

In this situation, you should train a Regularized model. Regularization is a penalization technique used during model training to avoid overfitting and improve generalization. It should help your Learning Curves to converge.
</details>

<details>
  <summary>What are the Support Vectors of an SVM model?</summary>

The support vectors are the observations upon which the model relies in order to place its separating hyperplane. The support vectors are located at the border between classes.
</details>

<details>
  <summary>What is the main risk with max margin classification?</summary>

The main risk is to overfit the model to the training data and not be able to generalize on unseen data points.
</details>

<details>
  <summary>What is the main difference between Ridge and Lasso regularization?

</summary>

* Increasing $\alpha$ in Ridge will only shrink parameters towards 0
* Increasing $\alpha$ in Lasso can shrink parameters to 0
</details>

<details>
  <summary>How would you ensure that you are not overfitting when tuning the hyperparameters of a model?</summary>

The workflow to avoid overfitting when model tuning is the following:

1. Holdout method on the entire dataset
2. Cross-validated model tuning on the train set
3. Final evaluation on the test set
</details>

<details>
  <summary>In the context of an SVM model, what is a kernel and what does it do? Name the kernels most often used with an SVM.</summary>

The 4 kernels allow the SVM to operate in a high-dimensional, non-linear feature space. They are:

* linear
* polynomial
* radial basis function
* sigmoid
</details>

<details>
  <summary>Where would the separating hyperplane of a max margin be located with respect to the support vector points?</summary>

The separating hyperplane of a max margin classifier would be located exactly half way between the support vectors of each class.
</details>

<details>
  <summary>What are the two most common Regularization techniques and their associated penalty terms?</summary>

* Ridge (L2) Regularization: its penalty is the **squared sum of each parameter**.
* Lasso (L1) Regularization: its penalty is the **absolute sum of each parameter**.
</details>

<details>
  <summary>Briefly explain how regularization prevents overfitting during model training.</summary>

Regularization reduces the complexity of the model and shrinks the learned estimates towards zero. In other words, it tunes the **loss function** by adding a **penalty term** which prevents excessive fluctuation of the coefficients.
</details>

<details>
  <summary>What are the two techniques for hyperparameter tuning?</summary>

Grid Search and Random Search are two hyperparameter tuning techniques.
</details>

No regularization at all! If $\alpha$ is zero, then the penalty term also equals zero.
</details>
