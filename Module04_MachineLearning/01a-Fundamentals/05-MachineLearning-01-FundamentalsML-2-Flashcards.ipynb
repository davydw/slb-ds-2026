<details>
  <summary>What is the standard ratio of train to test data when performing the Holdout Method?</summary>

When performing a holdout, the dataset is usually split in 75% of training data and 25% of test data.
</details>

<details>
  <summary>What are other words to name the X’s in a dataset?</summary>

There are many different words for the X’s:

* Variables
* Features
* Specificities
* Predictors
</details>

<details>
  <summary>What are bias and variance?</summary>

* Bias: The inability for an algorithm to learn the pattern of the dataset

8 Variance: Learning the training dataset so well to the point it can’t generalize unseen data
</details>

<details>
  <summary>What are learning curves and what are they used for?</summary>

Learning curves are used to diagnose overfitting and underfitting in a model. They visualize the effect of training set size on model performance.
</details>

<details>
  <summary>When reading learning curves, what are the two main signals of a performing model?

</summary>

* The score for both training and validation curves is relatively high
* Both curves have converged
</details>

<details>
  <summary>What is the difference between classification and regression?</summary>

* A classification task consists of predicting a label (e.g. hot or cold)
* A regression task consists of predicting a continuous value (temperature)
</details>

<details>
  <summary>In your own words, explain the “No Free Lunch” theory.</summary>

There is no one-size-fits-all model. Some models over-simplify, while others over-complicate a relationship between features and target. It’s up to us as data scientists to make assumptions about the data and evaluate reasonable models accordingly.
</details>

<details>
  <summary>What is unsupervised learning?</summary>

Unsupervised learning is the grouping and interpretation of data that does not have a target.
</details>

<details>
  <summary>What is supervised learning?</summary>

Supervised learning consists of modeling a relationship between input and output data.
</details>

<details>
  <summary>Classification tasks always have binary targets, true or false?</summary>

False! In classification, you can predict for as many different classes as you want.
</details>

<details>
  <summary>What is the K-fold Cross-validation algorithm and what is it used for?</summary>

Cross validation is used to generate less biased model performance estimates than the Holdout Method. Its steps are:

* The dataset is split into K number of folds
* Each fold is used both as training and testing across K number of splits (sub-models)
* The average score of all splits is the cross-validated score of the model
</details>

<details>
  <summary>If you were to predict whether it is going to rain or be sunny, would you use a Linear Regression or a Logistic Regression?</summary>

Predicting whether it is going to rain or be sunny is a classification task. Although it is called Logistic Regression, it is a classification model. You would therefore use Logistic Regression to solve the problem.
</details>

<details>
  <summary>What is the Holdout Method and why is it a key step for proper model evaluation?</summary>

The Holdout Method consists of splitting the dataset into two sets: train and test. It is used to evaluate a model’s ability to generalize, because it evaluates using data (the test set) that wasn’t seen during training.
</details>

<details>
  <summary>What are other words to name the y’s in a dataset?</summary>

Other words for y’s are:
* Target
* Class or Label in a classification task
</details>

<details>
  <summary>If you are working on a small dataset, how does the Holdout Method become problematic?

</summary>

The Holdout Method splits aside about 30% of the dataset for testing purposes. Those observations can therefore not be used for training. In the case of a small dataset, this could represent a significant information loss.
</details>

<details>
  <summary>Why is the score of a model obtained on a Holdout Method not a robust evaluation of performance?</summary>

When performing the Holdout Method, the data is randomly split into train and test sets. As such, different splits will create different results.
</details>

<details>
  <summary>What are bias and variance?</summary>

* Bias: The inability for an algorithm to learn the pattern of the dataset

* Variance: Learning the training dataset so well to the point it can’t generalize unseen data
</details>
