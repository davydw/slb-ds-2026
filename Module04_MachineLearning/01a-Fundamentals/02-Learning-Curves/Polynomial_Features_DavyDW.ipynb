But can we generalise the number of generated features for a higher number of features and a higher degree ? Probably. Let's go through it together.

````python
import numpy as np
import pandas as pd
pd.options.display.max_columns=None

from sklearn.preprocessing import PolynomialFeatures
````

````python
n = 1 # the number of rows does not matter, we wanna play with features and degrees
````

## 🥋 (1) Playing with degrees

### 🐣 Degree 1

There is nothing to generate: you have $p$ features: 
- $X_1, X_2, ..., X_p$

````python
p = 4 # features
d = 1 # degree

print(f"Playing with {p} feature(s)...")
print(f"...and d={d} degree(s)")
print("-"*25)

# CREATING THE DATAFRAME WITH N = 3 ROWS AND P FEATURES
X = pd.DataFrame(np.random.rand(n,p), columns = [f"X_{k+1}" for k in range(p)])
display(X)
print(f"🐣 We went from {p} feature(s)...")

# INSTANTIATING THE POLYNOMIAL FEATURES
poly = PolynomialFeatures(degree = d)
X_poly = pd.DataFrame(poly.fit_transform(X), columns = ["intercept"]+list(poly.get_feature_names_out()[1:]))
display(X_poly)
print(f"😈 ...to {X_poly.shape[1]-1} generated feature(s) + the intercept!")
````

### 🐥 Degree 2

* Squared features: 
  * $X_1^2, X_2^2, ... X_p^2 \rightarrow p $ squared features
* Products of features:
  * $(X_{\color{blue}{1}} \cdot X_2), (X_{\color{blue}{1}} \cdot X_3), ..., (X_{\color{blue}{1}} \cdot X_p) \rightarrow {\color{blue}{p-1}}$ generated features
  * $(X_{\color{purple}{2}} \cdot X_3), (X_{\color{purple}{2}} \cdot X_4), ..., (X_{\color{purple}{2}} \cdot X_p) \rightarrow {\color{purple}{p-2}}$ generated features
  * $(X_{\color{orange}{3}} \cdot X_4), (X_{\color{orange}{3}} \cdot X_5), ..., (X_{\color{orange}{3}} \cdot X_p) \rightarrow {\color{orange}{p-3}}$ generated features  
  * ...
  * $(X_{\color{red}{p-1}} \cdot X_p) \rightarrow {\color{red}{1}}$ generated features    

In total, `d=2` generated $p$ squared features and ${\color{red}{1}} + 2 + ... + {\color{orange}{p-3}} +{\color{purple}{p-2}} + {\color{blue}{p-1}} = \frac{(p-1)p}{2}$ products of features which combined exponents sum up to `d=2`

📚 Wanna know how to sum consecutive integers? [Brilliant.org - Wiki - Sum of n, n2 or n3](https://brilliant.org/wiki/sum-of-n-n2-or-n3/)

````python
p = 4 # features
d =2 # degree

print(f"Playing with {p} feature(s)...")
print(f"...and d={d} degree(s)")
print("-"*25)

# CREATING THE DATAFRAME WITH N = 3 ROWS AND P FEATURES
X = pd.DataFrame(np.random.rand(n,p), columns = [f"X{k+1}" for k in range(p)])
display(X)
print(f"🐣 We went from {p} feature(s)...")

# INSTANTIATING THE POLYNOMIAL FEATURES
poly = PolynomialFeatures(degree = d)
X_poly = pd.DataFrame(poly.fit_transform(X), columns = ["intercept"]+list(poly.get_feature_names_out()[1:]))
display(X_poly)
print(f"😈 ...to {X_poly.shape[1]-1} generated feature(s) + the intercept!")
````

### 🚀 Degree 3

* Cubes features: 
  * $X_1^3, X_2^3, ... X_p^3 \rightarrow p$ cubed features


* Products of features:
  * Multipling a squared feature by another one (because $2+1 =3$)
    * $(X_{\color{blue}{1}}^2 \cdot X_2), (X_{\color{blue}{1}}^2 \cdot X_3), ..., (X_{\color{blue}{1}}^2 \cdot X_p) \rightarrow {\color{blue}{p-1}}$ generated features
    * $(X_{\color{purple}{2}}^2 \cdot X_1), (X_{\color{purple}{2}}^2 \cdot X_3), ..., (X_{\color{purple}{2}}^2 \cdot X_p) \rightarrow {\color{purple}{p-1}}$ generated features
    * $(X_{\color{orange}{3}}^2 \cdot X_1), (X_{\color{orange}{3}}^2 \cdot X_2), ..., (X_{\color{orange}{3}}^2 \cdot X_p) \rightarrow {\color{orange}{p-1}}$ generated features  
    * ...
    * $(X_{\color{red}{p-1}}^2 \cdot X_1), (X_{\color{red}{p-1}}^2 \cdot X_2), ..., (X_{\color{red}{p-1}}^2 \cdot X_p) \rightarrow {\color{red}{p-1}}$ generated features  generated features    
    * $(X_{\color{green}{p}}^2 \cdot X_1), (X_{\color{green}{p}}^2 \cdot X_2), ..., (X_{\color{green}{p}}^2 \cdot X_{p-1}) \rightarrow {\color{green}{p-1}}$ generated features  generated features        
    * $\rightarrow p(p-1)$ generated features which follow the pattern $X_i^2 \cdot X_j$
  * Choosing $3$ distinct features among $p$ because $1+1+1 = 3$
    * $\rightarrow \binom{p}{3} = \frac{p!}{3!(p-3)!}$ generatures features which follow the pattern $X_i \cdot X_j \cdot X_k$

````python
p = 4 # features
d = 3 # degree

print(f"Playing with {p} feature(s)...")
print(f"...and d={d} degree(s)")
print("-"*25)

# CREATING THE DATAFRAME WITH N = 3 ROWS AND P FEATURES
X = pd.DataFrame(np.random.rand(n,p), columns = [f"X{k+1}" for k in range(p)])
display(X)
print(f"🐣 We went from {p} feature(s)...")

# INSTANTIATING THE POLYNOMIAL FEATURES
poly = PolynomialFeatures(degree = d)
X_poly = pd.DataFrame(poly.fit_transform(X), columns = ["intercept"]+list(poly.get_feature_names_out()[1:]))
display(X_poly)
print(f"😈 ...to {X_poly.shape[1]-1} generated feature(s) + the intercept!")
````

❓ How did we go from 4 features to 125 features ❓

Let's break down the `PolynomialFeatures(degrees = 4).fit_transform([X1, X2, X3, X4])` step by step.

#### 1️⃣ `X_poly_1`

````python
X_poly_1 = X_poly.iloc[:,1:5]
display(X_poly_1)
display(f"# polynomial features generated with degree 1 = {X_poly_1.shape[-1]}")
````

##### 💣 **Features to the power 1: $X_k^1$**
  * $X_1, X_2, ... X_p \rightarrow p = 4 $ features to the power of 1


* Original features $X_k$
  * $X_1, X_2, ..., X_p \rightarrow p = 4$ original features

#### 2️⃣ `X_poly_2`

````python
X_poly_2 = X_poly.iloc[:,5:15]
display(X_poly_2)
display(f"# polynomial features generated with degree 2 = {X_poly_2.shape[-1]}")
````

##### 💣 **Features to the power of 2: $X_k^2$** (squared features)
  * $X_1^2, X_2^2, ... X_p^2 \rightarrow p = 4 $ features to the power of 2


````python
mask_2_2_0 = [(("^2" in element) and (len(element)==(2+2))) for element in list(X_poly_2.columns)]
X_poly_2_2_0 = [feature for feature, b in zip(list(X_poly_2.columns), mask_2_2_0) if b]
display(X_poly_2_2_0)
print(f"len(X_poly_2_2_0) = {len(X_poly_2_2_0)}")
print(f"len(set(X_poly_2_2_0)) = {len(set(X_poly_2_2_0))}")
````

##### ✖️  Products of features $X_i \cdot X_j$

```quote
2 = 1 + 1
```

* $(X_{\color{blue}{1}} \cdot X_2), (X_{\color{blue}{1}} \cdot X_3), (X_{\color{blue}{1}} \cdot X_4) \rightarrow {\color{blue}{p-1}}$ generated features
* $(X_{\color{purple}{2}} \cdot X_3), (X_{\color{purple}{2}} \cdot X_4) \rightarrow {\color{purple}{p-2}}$ generated features
* $(X_{\color{orange}{3}} \cdot X_4) \rightarrow {\color{orange}{p-3}}$ generated features  
* ...
* $(X_{\color{red}{p-1}} \cdot X_p) \rightarrow {\color{red}{1}}$ generated features    
* ${\color{red}{1}} + 2 + ... + {\color{orange}{p-3}} +{\color{purple}{p-2}} + {\color{blue}{p-1}} = \frac{(p-1)p}{2} = \frac{(4-1) \cdot 4}{2} = \frac{(3) \cdot 4}{2} = 6$ products of features in the form $X_i \cdot X_j$



````python
mask_2_1_1 = [(not("^2" in element) and (len(element)==(2+1+2))) for element in list(X_poly_2.columns)]
X_poly_2_1_1 = [feature for feature, b in zip(list(X_poly_2.columns), mask_2_1_1) if b]
display(X_poly_2_1_1)
print(f"len(X_poly_2_1_1) = {len(X_poly_2_1_1)}")
print(f"len(set(X_poly_2_1_1)) = {len(set(X_poly_2_1_1))}")
````

##### 🌯 ***Wrapping up for 2*** 

````python
print(f"## len(X_poly_2_2_0) = {len(X_poly_2_2_0)}")
display(X_poly_2_2_0)
print(f"## len(X_poly_2_1_1) = {len(X_poly_2_1_1)}")
display(X_poly_2_1_1)
````

````python
total_counted_2 = len(X_poly_2_2_0) + len(X_poly_2_1_1)
total_counted_2
````

````python
display(f"# polynomial features generated with degree 2 = {X_poly_2.shape[-1]}")
````

## 🥷🏻 Generalisation 

### 🏋🏻‍♂️ Motivation

😱 If we have $p$ features and $d$ degrees, it will be very daunting to count the number of elements of the polynomial function with degree $d$...

### 🕵🏻‍♂️ Mission

🤔 What shall we do if we want to use $k$ features among the $p$ features that are available to generate a polynomial function with degree $d$ ? 


> 🎯 We need to create the following function:
> $$ X_{(1)}^{d_1} \cdot X_{(2)}^{d_2} ... X_{(k-1)}^{d_{k-1}} \cdot X_{(k)}^{d_k}  $$
> such that:
> $$d_{1} + d_{2} + ... + d_{k-1} + d_{k} = d$$

### 🔭 Taking a step back...

This is exactly what we did in our previous analysis with $d=5$ !

* 💣 $5 = 5 = d_{1}$
* ✖️ $5 = 4 + 1 = d_{1} + d_{2}$ or $(1+4)$
* ✖️ $5 = 3 + 2 = d_{1} + d_{2}$ or $(2+3)$
* ✖️ $5 = 3 + 1 + 1 = d_{1} + d_{2} + d_{3}$ 
  * or $(1+3+1)$
  * or $(1+1+3)$
* ✖️ $5 = 2 + 2 + 1 = d_{1} + d_{2} + d_{3}$
  * or $(2+1+2)$
  * or $(1+2+2)$
* ✖️ $5 = 1 + 1 + 1 + 2 = d_{1} + d_{2} + d_{3} + d_{4}$
  * or $(1+1+2+1)$ 
  * or $(1+2+1+1)$ 
  * or $(2+1+1+1)$
* ✖️ $5 = 1 + 1 + 1 + 1 + 1 = d_{1} + d_{2} + d_{3} + d_{4} + d_{5}$

### 🔢 The science of counting

##### Step 1 : choosing $k$ features among $p$

\begin{equation}
  {p\choose k} =
    \begin{cases}
      \frac{p!}{k!(p-k)!} & \text{if $0 \leq k \leq p$}\\
      0 & \text{otherwise}
    \end{cases}       
\end{equation}    



##### Step 2 : choosing $k$ degrees for the $k$ chosen features

For each feature $X_1, X_2, \dots, X_{k-1}, X_k$, you need to choose its respective degree $d_1, d_2, \dots, d_{k-1}, d_k$ such that  $d_{1} + d_{2} + \dots + d_{k-1} + d_{k} = d$

$ {d\choose d_1} \times {d-d_1\choose d_2} \times {d-(d_1+d_2)\choose d_3} \times \dots \times {d-(d_1 + d_2 + \dots + d_{k-2})\choose d_{k-1}} \times {d-(d_1 + d_2 + \dots + d_{k-1})\choose d_{k}}$

$= \frac{d!}{d_1!(d-d_1)!} \times \frac{(d-d_1)!}{d_2!(d-d_1-d_2)!} \times \frac{(d-d_1-d_2)!}{d_3!(d-d_1-d_2-d_3)!} \times \dots \times \frac{(d-d_1-d_2-\dots-d_{k-2})!}{d_{k-1}!d_k!} \times \frac{d_k!}{d_k!0!} $

$=\frac{d!}{d_1! d_2! ... d_k!}$

$= {d\choose{d_1,\dots,d_k}}$

> This number is called a **`multinomial coefficient`**!
> Read this article from Wikipedia 👉 [link](https://en.wikipedia.org/wiki/Multinomial_theorem).

##### Step 3 : How many combos with $k$ features, with a total degree $d$ ?

###### Multinomial Theorem - Generalised 

$$ \sum_{d_1 + \dots + d_k = d} {d\choose{d_1,\dots,d_k}} x_{(1)}^{d_1} \cdot x_{(2)}^{d_2} \dots x_{(k)}^{d_k}= (x_{(1)}+x_{(2)}+\dots+x_{(k)})^{d}$$

###### Multinomial Theorem with $x_{(1)} = x_{(2)} = \dots = x_{(k)} = 1$

$$ \sum_{d_1 + \dots + d_k = d} {d\choose{d_1,\dots,d_k}} = k^{d}$$

###### Number of multinomial coefficients

> The number of terms in a multinomial sum, $N_{d,p}$, is equal to the number of monomials of degree $d$ on the variables $X_1, X_2, ...X_k$:

$$N_{d,k} = {d + k -1 \choose k-1}$$

> Think about the [Stars and bars (combinatorics) strategy](https://en.wikipedia.org/wiki/Stars_and_bars_(combinatorics))

````python
def number_monomials(j, d):
  numerator = np.math.factorial(j+d-1)
  denominator = np.math.factorial(j)*np.math.factorial(d-1)
  return numerator / denominator
````
