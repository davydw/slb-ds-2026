# Fundamentals of Machine Learning

## 📚 Libraries

```python
# Data manipulation
import numpy as np
import pandas as pd
pd.options.display.max_columns = None

# Data Visualisation
import matplotlib.pyplot as plt

# Import the ML tools
from sklearn.model_selection import train_test_split

# Import the ML models
from sklearn.model_selection import cross_validate
```

## 💾 Dataset

```python
data = pd.read_csv("https://wagon-public-datasets.s3.amazonaws.com/Machine%20Learning%20Datasets/ML_Houses_dataset.csv")
data.head()
```

## 🎯 Defining your features and target

```python
# Feature.s - Pandas DataFrame
X = data[["GrLivArea"]]
```

```python
# Target - PandasSeries
y = data["SalePrice"]
```

## 🕵🏻‍♀️ Exploration

```python
# Plot Living area vs Sale price
plt.scatter(data['GrLivArea'], data['SalePrice'])

# Labels
plt.xlabel("Living area")
plt.ylabel("Sale price")

plt.show()
```

## 🤖 Modelling

```python
# Import the model
from sklearn.linear_model import LinearRegression

# Instanciate the model (💡 in Sklearn often called "estimator")
model = LinearRegression()

# Define X and y (if you haven't done it before)
X = data[['GrLivArea']]
y = data['SalePrice']

# Train the model on the data
model.fit(X, y)
```

## 👨🏻‍🔬 Model's attributes

"Fitting" a model means that the model is "learning" something. We will explain this later during the **`Under the hood** lesson. 

In the meantime, get used to carefully read the documentation!

```python
# View the model's slope (a)
model.coef_
```

```python
# View the model's intercept (b)
model.intercept_
```

## 🔮 Predicting

The trained model can be used to predict new data.

What is the predicted price of a 1000 $ft^2$ apartment ? 

```python
#  Predict on new data
model.predict([[1000]])
```

## 🔢 Scoring

Each Scikit-learn algorithm has a default scoring metric.

`LinearRegression` uses the **Coefficient of determination** ($R^2$) by default.

- $R^2$ represents the proportion of the variance of the target explained by the features.
- The score typically varies between 0 and 1
- The higher score the better the model

```python
# Evaluate the model's performance
model.score(X,y)
```

🚨 This score is not representative of the performance model because we trained the model on $(X,y)$ and evaluate on the same $(X,y)$

## ✂️ Holdout Method

```python
from sklearn.model_selection import train_test_split

# Define X and y
X = data[['GrLivArea']]
y = data['SalePrice']

# Split into Train/Test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Instantiate the model
model = LinearRegression()

# Train the model on the Training data
model.fit(X_train, y_train)

# Score the model on the Test data
model.score(X_test,y_test)
```

## 🏋🏻‍♂️ Model's robustness with Cross Validation

🤔 The randomness of the train_test_split will lead to different results from one run of the notebook to the other... from one colleague to the other...

<img src="https://miro.medium.com/max/1400/1*rgba1BIOUys7wQcXcL4U5A.png">

✅ Cross Validation to evaluate the robustness of a model

❌ Don't use it for predictions

```python
from sklearn.model_selection import cross_validate

# Instantiate model
model = LinearRegression()

# 5-Fold Cross validate model
cv_results = cross_validate(model,
                            X, y,
                            cv=5,
                            scoring = ["r2", "neg_mean_squared_error"])

# Scores
pd.DataFrame(cv_results)
```

```python
# For a R2, the higher the better
cv_results["test_r2"].mean()
```

```python
# For a MSE, the lower the better
# But Sklearn maximises "scoring", hence we look at the negative MSE and the higher the better
cv_results["test_neg_mean_squared_error"].mean()
```

👆 Which scoring method ?

- We will discuss it during **`Performance Metrics`**
- But you can already see that R2 in this situation is way more interpretable (usually between 0 and 1, explaining the variance of the target vs the features)
