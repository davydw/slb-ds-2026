# 05-MachineLearning-02-PrepareDataset-2-Flashcards

<details>
  <summary>When talking about a classification dataset, what is class imbalance?</summary>

Class imbalance means that there is not the same number of observations across targets.
</details>

<details>
  <summary>What is the one correct method for dealing with missing values in a dataset?</summary>

There is no one right method when it comes to missing values! The approach must be adapted to the nature of the missing values and the specific dataset.
</details>

<details>
  <summary>What is the curse of dimensionality?</summary>

As the number of features or dimensions grows, the amount of data we need to generalize accurately grows exponentially . As a rule of thumb, 10 rows are necessary to learn 1 feature. It is therefore important to reduce the dimensions of dataset as much as possible via feature selection.
</details>

<details>
  <summary>How can highly correlated features negatively affect a model?</summary>

Regression models assume each independent variable only has a relationship between itself and the dependent variable. If two independent variables are correlated, it would be impossible to change one without the other, which breaks the assumption of the regression model.
</details>

<details>
  <summary>Data leakage occurs when the same data points are present in both train and test sets. What can cause data leakage and how does it manifest itself?</summary>

If a dataset has duplicate observations, the same observations could end up both in both train and test sets. If that were the case, the performance of the model would be suspiciously high.
</details>

<details>
  <summary>What is scaling, and why is it important to scale features prior to modeling?</summary>

Scaling means transforming features into a common, smaller range.

* Features with large magnitudes can incorrectly outweigh features of small magnitudes
* Scaling to smaller magnitudes improves computational efficiency
* Scaling increases interpretability of feature coefficients
</details>

<details>
  <summary>Why would one choose to Robust scale rather than Standardize a feature?</summary>

In the case of a feature containing many outliers. Robust Scaling uses central tendency metrics that are less sensitive to outliers than the ones used by Standardization.
</details>

<details>
  <summary>What is the difference between Min Max scaling and Standardizing?</summary>

Min Max Scaling does not use central tendency measures. A feature will be transformed into a given range, commonly [0, 1], meaning that the minimum and maximum values of this feature are going to be 0 and 1, respectively.
</details>

<details>
  <summary>What is discretizing and in which situations can it be used?</summary>

Discretizing is the process of turning continuous data into discrete data according to sections (bins). It can be used to:

* Perform feature engineering
* Turn a regression task into a classification task
</details>

<details>
  <summary>What is a rule of thumb for preserving/dropping missing values?</summary>

* Greater than 30% of values missing: Potentially drop feature or row
* Less than 30% of values missing: impute a value that makes sense e.g. Median, mean, mode.
</details>

<details>
  <summary>List four reasons why feature selection is a key step of machine learning modeling.</summary>

* Faster training times
* Avoids the curse of dimensionality
* Reducing model complexity/ increases explainability
* Less redundant data means lower chance overfitting and making decisions based on noise
</details>

<details>
  <summary>What is the risk of encoding a multicategorical feature to a single numerical column? How should it be encoded instead?</summary>

Encoding a multicategorical feature to a single numerical column would give it a dimension of order. If the feature is not ordinal, it is preferable to One Hot Encode it.
</details>

<details>
  <summary>How does Standardization transform the data?</summary>

Standardization is a feature scaling technique that re-centers the data to 0 and the standard deviation to 1.
</details>

<details>
  <summary>You start investigating a feature, and the first thing you notice is that it has 50% missing values. You should remove the feature entirely. True or False?</summary>

False! Missing data does not necessarily mean no information! Make sure you investigate the meaning of the missing values first.
</details>

<details>
  <summary>What is the strategy to avoid data leakage when balancing a dataset?</summary>

In order to avoid data leakage, balancing of the classes is performed on the training set only.
</details>

<details>
  <summary>Unwanted duplicated observations can discredit the performance evaluation of a model. Why?</summary>

In order to evaluate a model’s ability to generalize, it is crucial that the data in the test set remains unseen by the algorithm throughout training. If duplicates exist in both training and testing this can cause unreliable scores.
</details>

<details>
  <summary>Why is it important to balance the targets of a classification dataset prior to performing Machine Learning?</summary>

Certain Machine Learning models are sensitive to the proportion of targets and are unable to cope with imbalance.
</details>

<details>
  <summary>How does the feature permutation algorithm work for feature selection?</summary>

Feature permutation evaluates the importance of each feature in predicting the target. It does so by observing how the performance of the model is affected when the feature is randomly shuffled.
</details>

<details>
  <summary>When detecting missing values, what is the danger of Python’s isnull() function?</summary>

Python’s `isnull()` only detects np.nans, and will miss any other representation of missing values.
</details>
