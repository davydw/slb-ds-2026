## Dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_wine

wine = load_wine(as_frame=True)
X = wine.data
y = wine.target
wine_features = X.columns
```

```python
X.shape
```

```python
X.head()
```

```python
# y.value_counts()
```

## Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().set_output(transform="pandas")
scaler.fit(X)
X_scaled = scaler.transform(X)
X_scaled
```

```python
# X.columns
```

```python
# scaler.get_feature_names_out()
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(X.corr(), annot = True, cmap = "PuRd");
```

## Principal Component Analysis

### Finding the components

```python
from sklearn.decomposition import PCA

# Instantiate the PCA
pca = PCA().set_output(transform="pandas")

# Find the components
pca.fit(X_scaled)
```

```python
X_scaled.shape
```

```python
# Show the components
# (n_components, n_features)
pca.components_
```

```python
pca.components_.shape
```

```python
components = pd.DataFrame(pca.components_ ).T
components.columns = [f'PC{i+1}' for i in range(0, 13)]
components.index = X.columns
components
```

```python
X_scaled
```

### Projecting X_scaled into a new space

```python
X_proj = pca.transform(X_scaled) # np.dot(X_Scaled, components)
X_proj
```

```python
pd.DataFrame(np.dot(X_scaled, components), columns= [f'PC{i}' for i in range(0, 13)])
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(X_proj.corr(), cmap = "PuRd");
```

### Finding the ideal number of components to be kept

```python
# Variance for each component
X_proj.std()**2
```

```python
total_variance = (X_proj.std()**2).sum()
total_variance
```

```python
# Share of variance for each component
X_proj.std()**2/total_variance
```

```python
# We can show the share of variance
pca.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
plt.plot(pca.explained_variance_ratio_)
plt.xlabel('Principal Component'); plt.ylabel('% explained variance');
```

```python
import numpy as np
np.cumsum(pca.explained_variance_ratio_)
```

```python
plt.plot(range(1,13+1),np.cumsum(pca.explained_variance_ratio_))
plt.grid()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate
```

```python
model1 = LogisticRegression()
cv1 = cross_validate(model1, X_scaled, y) # 13 FEATURES
cv1['test_score'].mean()
```

```python
model2 = LogisticRegression()
cv2 = cross_validate(model2, X_proj.loc[:,["pca0", "pca1", "pca2"]], y) # 3 PRINCIPAL COMPONENTS
cv2['test_score'].mean()
```

## KMeans

```python
from sklearn.cluster import KMeans

# Instantiate K-Means
# by default n_clusters = 8
km = KMeans(n_clusters=3) 
km.fit(X_proj)
```

```python
# what are the coordinates of my centroids ?
my_centroids = pd.DataFrame(km.cluster_centers_)
my_centroids.columns = X_proj.columns
my_centroids
```

```python
km.cluster_centers_.shape
```

```python
X.shape
```

```python
km.labels_
```

```python
plt.scatter(X.iloc[:,0], X.iloc[:,1])
```

```python
# LET'S VISUALIZE OUR 180-ISH BOTTLES OF WINE IN THE SPACE OF ORTHOGONAL COMPONENTS
plt.scatter(X_proj.iloc[:,0], X_proj.iloc[:,1], c = km.labels_)

# 0 for pca 0
# 1 for pca 1

# CENTROID 0
plt.scatter(my_centroids.iloc[0,0], my_centroids.iloc[0,1], c="purple", s=500)
# CENTROID 1
plt.scatter(my_centroids.iloc[1,0], my_centroids.iloc[1,1], c="green", s=500)
# CENTROID 2
plt.scatter(my_centroids.iloc[2,0], my_centroids.iloc[2,1], c="red", s=500)

# LET'S GO
plt.grid()
plt.show()
```

```python
# How do I choose the number of clusters ?
inertias = []
list_of_k = range(1, 10)

for k in list_of_k:
  temp_km = KMeans(n_clusters = k)
  temp_km.fit(X_proj)
  inertias.append(temp_km.inertia_)
```

```python
plt.plot(list_of_k, inertias)
plt.scatter(3, inertias[3-1], c="green", s=500)
plt.grid()
```
