## 📚 Libraries

```python
# DATA MANIP
import pandas as pd
import numpy as np

# DATA VIZ
import seaborn as sns
import matplotlib.pyplot as plt

# MACHINE LEARNING

### HOLD OUT METHOD / CROSS VAL
from sklearn.model_selection import train_test_split, cross_validate

### PREPROCESSING
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

### MODELS
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

### METRICS
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import classification_report
```

```python
data = sns.load_dataset("titanic").drop(columns=["alive", "embarked", "class", "who"])
data.head()
```

## 🎯 Target

```python
y = data["survived"]
X = data.drop(columns=["survived"])
```

## ✂️ Train-test-split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y) #test_size = 0.25)
```

```python
X.shape, X_train.shape, X_test.shape
```

## 🧼 Real cleaning

```python
X_train
```

```python
X_train.dtypes.value_counts()
```

```python
# X_train_num  = X_train[["pclass", "age", "sibsp", "parch", "fare"]]
# X_train_num = X_train.select_dtypes(include = ["int64", "float64"])
# X_train_num = X_train.select_dtypes(include = np.number)


X_train_num = X_train.select_dtypes(include = "number")
X_test_num = X_test.select_dtypes(include = "number")

X_train_cat = X_train.select_dtypes(exclude = "number")
X_test_cat = X_test.select_dtypes(exclude = "number")
```

### 🔢 Scaler

```python
imputer_num = SimpleImputer(strategy = "mean")
imputer_num.fit(X_train_num[["age"]])

X_train_num["age"] = imputer_num.transform(X_train_num[["age"]])
X_test_num["age"] = imputer_num.transform(X_test_num[["age"]])
```

```python
scaler = StandardScaler().set_output(transform="pandas")

scaler.fit(X_train_num)

X_train_num_scaled = scaler.transform(X_train_num)
X_test_num_scaled = scaler.transform(X_test_num)
```

```python
X_train_num.head()
```

```python
X_train_num_scaled.head()
```

```python
# scaler.get_feature_names_out()
```

```python
# pd.DataFrame(X_train_num_scaled, columns = scaler.get_feature_names_out())
```

### 🔡 Encoder

```python
imputer_cat = SimpleImputer(strategy="most_frequent")

imputer_cat.fit(X_train_cat[["embark_town"]])
```

```python
X_train_cat["embark_town"] = imputer_cat.transform(X_train_cat[["embark_town"]])
X_test_cat["embark_town"] = imputer_cat.transform(X_test_cat[["embark_town"]])
```

```python
X_train_cat.isnull().sum()
```

```python
X_test_cat.isnull().sum()
```

**Encoding**

```python
encoder = OneHotEncoder(
    sparse_output=False,
    drop="if_binary",
    handle_unknown="ignore"
).set_output(transform="pandas")

encoder.fit(X_train_cat)

X_train_cat_encoded = encoder.transform(X_train_cat)
X_test_cat_encoded = encoder.transform(X_test_cat)
```

```python
X_train_cat.head()
```

```python
X_train_cat_encoded.head()
```

```python
# pd.DataFrame(X_train_cat_encoded, columns = encoder.get_feature_names_out())
```

### 🔢 + 🔠 All together

```python
X_train_preprocessed = pd.concat([X_train_num_scaled, X_train_cat_encoded], axis="columns")
X_test_preprocessed = pd.concat([X_test_num_scaled, X_test_cat_encoded], axis="columns")
```

```python
X_train_preprocessed.head()
```

## 🤖 Modelling

### Logistic Regression

```python
model = LogisticRegression(max_iter = 300)

model.fit(X_train_preprocessed, y_train)
```

```python
y_pred = model.predict(X_test_preprocessed)
```

### KneigbhorsClassifier

```python
def classif_2062(model, X_train, X_test, y_train, y_test):
    #### MODEL INSTANTIATION
    model = model
    #### MODEL TRAINING
    model.fit(X_train, y_train)
    #### MODEL PREDICTIONS ON THE TEST SET
    y_pred = model.predict(X_test)
    #### CONFUSION MATRIX
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.show()
    #### CLASSIFICATION REPORT
    print(classification_report(y_test, y_pred))
    return model
```

```python
classif_2062(KNeighborsClassifier(), 
             X_train_preprocessed, 
             X_test_preprocessed, 
             y_train, y_test)
```
