# 05-MachineLearning-03-PerformanceMetrics-2-Flashcards

<details>
  <summary>When tuning a KNN, what is the risk of setting a value of K that is too low?</summary>

Low k values could lead to overfitting the training set. The model needs a certain number of observations to base its predictions on. Usually (but not always!), the ideal K lies between 5 and 10.
</details>

<details>
  <summary>Is K-Nearest Neighbors (KNN) a classification or regression model?</summary>

K-Nearest Neighbors (KNN) is a Machine Learning algorithm that can perform both regression and classification tasks.
</details>

<details>
  <summary>What are the two metrics that are represented in the ROC curve?</summary>

The ROC curve represents the tradeoff between the True Positive rate and the False Positive rate. The True Positive rate is none other than the recall.
</details>

<details>
  <summary>Why do we use baseline scores?</summary>

Baseline scores are used as reference points when starting the iterative model lifecycle. They ensure we move in the right direction from the very first iteration.
</details>

<details>
  <summary>If you were training a model to detect Covid-19 infected patients, which performance metric would you look at?</summary>

When it is important to identify as many occurences of a class as possible, such as Covid-19 infected people, the performance metric to focus on is the recall.
</details>

<details>
  <summary>What is a confusion matrix, and what are its elements?</summary>

A table in which the count of correct and incorrect predictions are broken down. It is the basis of most Classification metrics and is made up of:

* True positives
* False positives
* True negatives
* False negatives
</details>

<details>
  <summary>What is the main parameter of the KNN model?</summary>

The main parameter of the KNN model is the number of neighbors.
</details>

<details>
  <summary>If you were training a classification model to detect potential investment opportunities and wanted to limit your risk, which metric would you focus on?</summary>

When it is important to be right when identifying a specific class (in this case, a good investment), the metric to look at is precision.
</details>

<details>
  <summary>What element of a probabilistic model needs to be adjusted to enhance the precision or recall?</summary>

The decision threshold, which is 0.5 by default, can be adjusted to enhance the recall or precision scores of a model.
</details>

<details>
  <summary>What is an alternative to the accuracy score when dealing with an imbalanced dataset?</summary>

The f1 score is similar to the accuracy, but is robust to class imbalance.
</details>

<details>
  <summary>How do KNN regressors make their final prediction?</summary>

A KNN regressor will predict the mean value of the K points that are the closest to the one it is trying to predict.
</details>

<details>
  <summary>Why might Normalizing features enhance the performance of KNN’s over Standardizing?</summary>

KNN’s and distance based algorithms are very sensitive to the scale of features. As such, ensuring an exact common range between each feature (normalizing) will often produce better results.
</details>

<details>
  <summary>Give an example of baseline scores for classification and regression.</summary>

* **Classification**: a random prediction of the target classes
* **Regression**: a central tendency measure like the mean or median of the target
</details>

<details>
  <summary>Which regression metric would you use to penalize large errors and have the error be in the unit of the target?</summary>

The Root Mean Squared error. Squaring will penalize large errors by blowing them out of proportion. Taking the square root will then transform the error back to the unit of the target.
</details>

<details>
  <summary>Which classification metrics measures the ability of a model to detect occurences of a class? What is its formula?</summary>

The recall is the ability of a model to identify occurences of a class = TP / (TP + FN)
</details>

<details>
  <summary>Which regression metric has scores typically ranging between 0 and 1 (though sometimes produces negative scores!)? What does it represent?

</summary>

* The coefficient of determination R2 measures the proportion of the variance in the dependent variable that is predictable from the independent variable(s). 
* N.B. if R2 becomes negative the predictions are worse than the mean.
</details>
