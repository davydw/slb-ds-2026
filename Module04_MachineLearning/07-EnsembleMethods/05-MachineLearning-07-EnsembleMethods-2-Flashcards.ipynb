# 05-MachineLearning-07-EnsembleMethods-2-Flashcards.ipynb

<details>
  <summary>If you are training an ensemble of bootstrap aggregated decision trees, what model are you actually using?</summary>

A Random Forest model! Random Forest is the bagged (bootstrap aggregated) version of decision trees.
</details>

<details>
  <summary>When training Decision Tree or Random Forest models, features do not have to be scaled. True or False?</summary>

True! One advantage of decision trees and random forests is that they are not affected by the varying scales of features.
</details>

<details>
  <summary>You are training an algorithm that has high bias. You’ve tried everything to lower the bias and nothing worked. The one thing you haven’t tried is Ensemble methods. Which technique do you go for</summary>

Boosting! The main objective of Boosting is to reduce bias.
</details>

<details>
  <summary>The Boosting ensemble technique can only be applied to decision trees. True or False?</summary>

False again! Both bagging and boosting can be applied to any algorithm!!
</details>

<details>
  <summary>The Bagging ensemble technique can only be applied to decision trees. True or False?</summary>

False! Bagging works well with decision trees (Random Forest), but can be applied to any algorithm!
</details>

<details>
  <summary>What measure does the hierarchical structure of decision trees depend on?</summary>

The hierarchical structure of decision trees is based on the Gini index. The Gini index is a measure that indicates the ability of each feature to separate the data. We say that it calculates the impurity.
</details>

<details>
  <summary>In bagging, all weak learners have the same weight in the final prediction votes. True or False?</summary>

True! In bagging, all weak learners are attributed the same importance in the final vote.
In boosting however, the best weak learners are attributed more weight in the final vote.
</details>

<details>
  <summary>To which family of ensemble methods do Bagging and Boosting belong to? Sequential or Parallel?</summary>

* Bagging is a parallel ensemble method.
* Boosting is a sequential ensemble method.
</details>

<details>
  <summary>How would you prevent your decision tree model from overfitting?</summary>

Decision trees absolutely need to be tuned to avoid overfitting! This can be done by optimizing the leaf size and depth parameters of the model.
</details>

<details>
  <summary>Explain how outliers can have a negative effect on boosting.</summary>

In a way, Boosting consists of spending more time trying to learn observations that are hard to predict. If that observation is an outlier, there is a good chance the model will never be able to learn to predict it. In that case, the boosted algorithm can be stuck in a loop trying to solve the impossible.
</details>

<details>
  <summary>Random forests solve an issue specific to decision trees. What is that issue?</summary>

Decision trees are very sensitive to the data they are trained on and have a high variance. Bagging decision trees (Random Forest) reduces that variance.
</details>

<details>
  <summary>If you are “ensembling” multiple models for a classification task, how would you combine the predictions of the different models?</summary>

In the case of classification, you can perform a majority vote across the predictions of the models.
</details>

<details>
  <summary>If you are “ensembling” multiple models for a regression task, how would you combine the predictions of the different models?</summary>

In a regression task, the models are aggregated by averaging their predictions.


</details>
