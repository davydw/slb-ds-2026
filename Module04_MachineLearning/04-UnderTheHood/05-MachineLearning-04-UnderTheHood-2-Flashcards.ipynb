# 05-MachineLearning-04-UnderTheHood-2-Flashcards

<details>
  <summary>What is the learning rate in Gradient Descent?</summary>

It is the value that, multiplied with the gradient of the loss function, determines how much the parameters are updated from one iteration to the next.
</details>

<details>
  <summary>Name two possible Loss functions for a linear regression model.</summary>

* The Mean Squared Error
* The Mean Absolute Error
</details>

<details>
  <summary>What is the main difference between using Mean Squared Error and Mean Absolute Error losses?</summary>

* Mean Squared Error loss penalizes big errors more than small errors.
8 Mean Absolute Error loss penalizes all errors equally
The predictions of a model optimized on an MSE loss will be more centered around the true values than a model optimized on the Mean Absolute Error.
</details>

<details>
  <summary>Remind me the steps of a Gradient Descent epoch.</summary>

1. Randomly initialize parameter value
2. Compute Loss function derivative at that point
3. Update the parameter value with the step size
4. Go back to step 2 with updated parameter value
5. Repeat until model converges to minimum loss
</details>

<details>
  <summary>What is the Mini-Batch Gradient Descent?</summary>

Parameters are updated after computing the gradient of error with respect to a subset of the training set
</details>

<details>
  <summary>In Gradient Descent, what is used as an indicator of minimum loss? What value of that indicator would suggest a minimum loss?</summary>

The slope (or gradient) of the Loss function is the indicator of minimum loss. As the slope approaches zero, the loss approaches its minimum.
</details>

<details>
  <summary>How is the slope of the Loss function computed?</summary>

The slope equals the value which set the partial derivative with respect to the parameter of interest to zero.
</details>

<details>
  <summary>What could happen if the learning rate is too small?</summary>

If the learning rate is too small, the Gradient Descent could get stuck in a local minimum. If it were to converge, it would require many more epochs than with a higher learning rate.
</details>

<details>
  <summary>You are training a model to predict the amount of time a diver can go on his compressed air tank. Which Loss function do you optimize your model on?</summary>

In this particular case, big errors could cost a diver their life. As such, you should optimize your model on a Mean Squared Error loss to minimize large errors.
</details>

<details>
  <summary>You train an SGDClassifier() and set the parameter loss='log'. What type of model are you using and how is the model being optimized?</summary>

The log loss corresponds to a Logistic Regression model which in this case is optimized by Stochastic Gradient Descent.
</details>

<details>
  <summary>What is the difference between the loss and the performance metric?</summary>

* The loss metric is used during training to optimize the model
* The performance metric is used after training to evaluate the performance of the model
</details>

<details>
  <summary>What is the difference between stochastic gradient descent and batch gradient descent?</summary>

* Batch Gradient Descent: Parameters are updated after computing the gradient of error with respect to the batch, which is only a subset of the training data
* Stochastic Gradient Descent: Parameters are updated after computing the gradient of error with respect to a single training example
</details>

<details>
  <summary>Gradient Descent consists of minimizing a function. What is that function called?</summary>

The function Gradient Descent is trying to minimize is called the Loss function. By minimizing the Loss function, the model will be as close as possible to the true value it is trying to predict.
</details>

<details>
  <summary>The loss and the performance metric can sometimes be the same. True or False?</summary>

The Loss function is used to optimize your model. This is the function that will get minimized by the optimizer.
A performance metric is used to judge the performance of your model. The performance metric has nothing to do with the optimization process.
The same methodology can be used for both cases so… Yes! For example, the Mean Squared Error (MSE) can be used as the loss and performance metric.
This is often done in Linear Regression, where the model is both optimized (loss metric) and evaluated (performance metric) by the same method, the MSE.
Note that we define a well-performing model as one with a low MSE (they have an inverse relationship) and therefore put a minus in front, performance = -MSE. A high loss, on the other hand, corresponds directly to a high MSE, loss = MSE.
</details>
