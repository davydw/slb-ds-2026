2. Use the **requests** library to get the HTML file of the web page.

4. Create a new **DataFrame** and fill it with the data.

<details>
  <summary>What are regular expressions?</summary>

Regular expressions are essentially a tiny, highly specialized text-processing tool used within many programming languages to find `str`ings matching a pattern inside a chain of characters. It is embedded inside Python and made available through the `re` module.
