# Scrape data for One player



Retrieve the name of the first player (Traditional method).


We will use HTML tags to navigate through our HTML doc, as we would do normally with BeautifulSoup.


```python
# We can achieve this by using the find_element() function of the Selenium WebDriver. Here we navigated through the HTML by using the class names inside the HTML tags.

driver.find_element(By.CLASS_NAME,"odd").find_element(By.CLASS_NAME,"inline-table").find_element(By.CLASS_NAME,"hauptlink").text
```

Retrieve the name of the first player (Advanced method).

We will use the XPATH of a specific element in our HTML doc to access it directly.

↪️ When hovering over an HTML section in the HTML doc : **Right Click > Copy > Copy XPath**




```python
#The XPATH method does the same thing but in a much quicker way and with less code to produce

name=driver.find_element(By.XPATH,'//*[@id="yw1"]/table/tbody/tr[1]/td[2]/table/tbody/tr[1]/td[2]/a').text
name
```

Retrieve the position of the first player



```python
pos=driver.find_element(By.XPATH,'//*[@id="yw1"]/table/tbody/tr[1]/td[2]/table/tbody/tr[2]/td').text
pos
```

Retrieve the age of the first player



```python
age=int(driver.find_element(By.XPATH,'//*[@id="yw1"]/table/tbody/tr[1]/td[3]').text)
age
```

Retrieve the club of the first player




```python
#There is no text inside the HTML tag so we cannot use the .text here.
# We can however use the get_attribute() method to retrieve the value associated to the title attribute, whoch contains what we are looking for.

club=driver.find_element(By.XPATH,'//*[@id="yw1"]/table/tbody/tr[1]/td[5]/a/img').get_attribute("title")
club
```

Retrieve the market value of the first player



```python
mv=float(driver.find_element(By.XPATH,'//*[@id="yw1"]/table/tbody/tr[1]/td[6]/a').text[1:-1])
mv
```

Retrieve the nationalities of the first player.

We might be in trouble since some players have more than one nationality, resulting in different code structures in our HTML !

We can work around that by combining the Traditional and the Advanced methods of scraping (more details below).



```python
#First, we retrieve the Section containing all the nationalities for one player, and store it in a variable called nat_box.

nat_box=driver.find_element(By.XPATH,f'//*[@id="yw1"]/table/tbody/tr[1]/td[4]')

#Then we use the find_elements() function (same as find_element but to retrieve a list of matching elements). We will use the class "flaggenrahmen" shared by all the nationalities

nationalities=nat_box.find_elements(By.CLASS_NAME,"flaggenrahmen")

# We can now initiate a loop to retrieve for each nationality the "title" attribute, and append its value to a list called nat_list.

nat_list=[]
for nat in nationalities:
    nat_list.append(nat.get_attribute("title"))

# We can join every element of nat_list into a single string by using the .join() method
nationalities=", ".join(nat_list)
nationalities
```

# Scraping data for all players on the 1st page


Print the name of the 3 first players. Is there a pattern in the XPATH we can exploit?


```python
print(driver.find_element(By.XPATH,'//*[@id="yw1"]/table/tbody/tr[1]/td[2]/table/tbody/tr[1]/td[2]/a').text)
print(driver.find_element(By.XPATH,'//*[@id="yw1"]/table/tbody/tr[2]/td[2]/table/tbody/tr[1]/td[2]/a').text)
print(driver.find_element(By.XPATH,'//*[@id="yw1"]/table/tbody/tr[3]/td[2]/table/tbody/tr[1]/td[2]/a').text)

```

Print the age of the 3 first players. Do we see the same pattern ?


```python
print(driver.find_element(By.XPATH,'//*[@id="yw1"]/table/tbody/tr[1]/td[3]').text)
print(driver.find_element(By.XPATH,'//*[@id="yw1"]/table/tbody/tr[2]/td[3]').text)
print(driver.find_element(By.XPATH,'//*[@id="yw1"]/table/tbody/tr[3]/td[3]').text)

```

We can exploit it to loop and retrieve the data for all players of the page!


```python
#We can re-use what was done previously by iterating on the 25 players of the first page

players=[]

for nb in range(1,26):

    name=driver.find_element(By.XPATH,f'//*[@id="yw1"]/table/tbody/tr[{nb}]/td[2]/table/tbody/tr[1]/td[2]/a').text
    pos=driver.find_element(By.XPATH,f'//*[@id="yw1"]/table/tbody/tr[{nb}]/td[2]/table/tbody/tr[2]/td').text
    age=int(driver.find_element(By.XPATH,f'//*[@id="yw1"]/table/tbody/tr[{nb}]/td[3]').text)
    club=driver.find_element(By.XPATH,f'//*[@id="yw1"]/table/tbody/tr[{nb}]/td[5]/a/img').get_attribute("title")
    mv=float(driver.find_element(By.XPATH,f'//*[@id="yw1"]/table/tbody/tr[{nb}]/td[6]/a').text[1:-1])
    nat_box=driver.find_element(By.XPATH,f'//*[@id="yw1"]/table/tbody/tr[{nb}]/td[4]')
    nationalities=nat_box.find_elements(By.CLASS_NAME,"flaggenrahmen")
    nat_list=[]
    for nat in nationalities:
        nat_list.append(nat.get_attribute("title"))

    nationalities=", ".join(nat_list)

    player=[name,pos,age,club,nationalities,mv]
    players.append(player)
```

```python
#We can now store the data retrieved inside a DataFrame

pd.DataFrame(players,columns=["Name","Position","Age","Club","Market Value","Nationalities"])

```

# Scraping data from multiple pages


Let's access the second page


```python
driver.find_element(By.XPATH,'//*[@id="yw1"]/div[2]/ul/li[11]/a').click()
```

Now we have everything that we need ! Let's scrape the 500 most expensive football players

```python
driver.quit()
```

```python
players=[]

#the following lines impose some waiting time to the driver. This waiting time is necessary because the website is coded dynamically, therefore we need to wait for the "Next Page" button to be clickable.

wait = WebDriverWait(driver, 10)
next_button = wait.until(
    EC.element_to_be_clickable((By.XPATH, "//ul[@class='tm-pagination']/li[last()-1]/a")),
    message="Next button not clickable")

#In order to reach the "Next Page" button, we have to scroll down so that the Selenium Webdriver can click on it
target_loc=next_button.location
driver.execute_script("window.scrollTo(arguments[0], arguments[1]);", target_loc['x'], target_loc['y']-700)


#25 players per page
nb_players=500
max_page=int(nb_players/25)

for page in range(max_page):
    for num in range(1,26):
        age=int(driver.find_element(By.XPATH,f'//*[@id="yw1"]/table/tbody/tr[{num}]/td[3]'))
        name=driver.find_element(By.XPATH,f'//*[@id="yw1"]/table/tbody/tr[{num}]/td[2]/table/tbody/tr[1]/td[2]/a')
        position=driver.find_element(By.XPATH,f'//*[@id="yw1"]/table/tbody/tr[{num}]/td[2]/table/tbody/tr[2]/td')
        club=driver.find_element(By.XPATH,f'//*[@id="yw1"]/table/tbody/tr[{num}]/td[5]/a/img')
        value=driver.find_element(By.XPATH,f'//*[@id="yw1"]/table/tbody/tr[{num}]/td[6]/a')

        nat_box=driver.find_element(By.XPATH,f'//*[@id="yw1"]/table/tbody/tr[{num}]/td[4]')
        nationalities=nat_box.find_elements(By.CLASS_NAME,"flaggenrahmen")
        nat_list=[]
        for nat in nationalities:
            nat_list.append(nat.get_attribute("title"))

        player=[
            name.text,
            position.text,int(age.text),
            ", ".join(nat_list),
            club.get_attribute("title"),
            float(value.text[1:-1])
        ]
        players.append(player)

    if page <max_page-1:
        next_button.click()
    # Wait for the next button to become clickable again
        wait.until(EC.staleness_of(next_button))
        next_button = wait.until(
            EC.element_to_be_clickable((By.XPATH, "//ul[@class='tm-pagination']/li[last()-1]/a")),
            message="Next button not clickable")
    else:
        break

driver.quit()
```

```python
# Let's load our data into a DataFrame

players_df=pd.DataFrame(players,columns=["Name","Position","Age","Nationalities","Club","Market_value_(m€)"])
players_df
```

```python
# We can now perform some aggregations

players_df.groupby("Club").sum()[["Market_value_(m€)"]].sort_values("Market_value_(m€)",ascending=False)
```

```python
#We can export the data by using various functions like pandas.to_csv() , pandas.to_gbq() etc.

players_df.to_csv("players_mv_2024.csv")
```

# Bonus - What if we wanted to scrape each player's page ?





If we have a look at the player's name section, we can see that there is a link to the player's page embedded.


```python
driver.find_element(By.XPATH,'//*[@id="yw1"]/table/tbody/tr[1]/td[2]/table/tbody/tr[1]/td[2]/a').get_attribute("href")
```

We cannot access it by using the .click() method though.
We can however store it in a new list of urls to scrape later !


```python
urls=[]
for nb in range(1,26):

    url = driver.find_element(By.XPATH,f'//*[@id="yw1"]/table/tbody/tr[{nb}]/td[2]/table/tbody/tr[1]/td[2]/a').get_attribute("href")
    urls.append(url)

urls
```

We can now use these newly acquired urls to move our Webdriver to new pages and scrape further! 🚀

```python
first_url=urls[0]
driver.get(first_url)
```
