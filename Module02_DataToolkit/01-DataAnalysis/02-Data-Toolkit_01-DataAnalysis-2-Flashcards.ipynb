<details>
  <summary>What are the steps needed to group (=GROUP BY in SQL) the rows of a DataFrame?</b></summary>


1. **Split** the DataFrame into groups defined by a chosen key (=GROUP BY some_key in SQL)
2. **Apply an aggregative function** (sum, mean, etc.) on each group
3. Chain the previous operations and store the result into one new DataFrame

E.g. 
```python
avg_data = data.groupby(by='age').mean()
```
</details>
