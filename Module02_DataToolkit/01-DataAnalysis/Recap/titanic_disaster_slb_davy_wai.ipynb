## Children

The former analysis did not take into account ages. We want to differentiate between a child and an adult and see how *survival rates* are affected.

❓ Use boolean indexing to create a `children_df` containing only rows of child passengers

```python
adult_threshold = 18
```

```python
children_mask = titanic_df["Age"] < adult_threshold
```

```python
children_df = titanic_df[children_mask]
children_df.head()
```

```python
# SHOW A RANDOM ROW OF THE DATAFRAME
children_df.sample()
```

```python
children_df.isna().sum().sort_values(ascending=False)
```

```python
# SHOW 20 RANDOM PASSENGERS OF THE DATAFRAME
children_df.sample(20)
```
