## Data selection 😎

```python
# lets build a 2D array from a list of list
data_list = [
    [ 0,  1,  2,  3,  4],
    [10, 11, 12, 13, 14],
    [20, 21, 22, 23, 24],
    [30, 31, 32, 33, 34],
    [40, 41, 42, 43, 44],
]
data_np = np.array(data_list)
data_np
```

### Slicing within a row

🤔 How can we select specific values from a **ROW** ?


```python
# Pure python
data_list[2][1:4]
```

#### NumPy

```python
# NumPy
data_np[2, 1:4] # data_np[row(s), column(s)]
```

### Slicing within a column

🤔 How can we select specific values from a **COLUMN** ?


#### Pure Python

```python
# Pure python
selection = []
for index, row in enumerate(data_list):
    if index > 0:
        selection.append(row[4])
selection # We could also have used a list comprehension for less lines
```

#### Numpy

```python
# NumPy
data_np[1:, 4] # `1:` means from line 1 until the end.
```

### General Syntax

```python
A[1:7:2]
```

## Vectorized operations ⚡️

Let's compute the sum row-by-row (8 additions) to create a 1D vector

```python
my_list = [
    [6, 5],
    [1, 3],
    [5, 6],
    [1, 4],
    [3, 7],
    [5, 8],
    [3, 5],
    [8, 4],
]
```

### Python Way

```python
# Python way
sums = []
for row in my_list:
    sums.append(row[0] + row[1]) # standard integer "+" operator
sums
```

### The NumPy way

```python
A = np.array(my_list)
my_sum = A[:,0] + A[:,1] # Vectorial "+" operator
my_sum
```

## Intro

### Pandas Series

* Pandas equivalent to numpy **1D ndarray** (accept same methods)
* with an additional **index**
* with support for **multiple data types**

👉 [pandas.Series](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.html)

```python
s = pd.Series(data=[1,2,'three'], index=['id1', 'id2', 'id3'])
s = pd.Series({'id1': 1, 'id2': 2, 'id3': 'three'})
s
```

### Pandas DataFrames

Pandas equivalent of a NumPy 2D ndarray :

   * with additional labels on both axes (row and columns)
   * with multiple data types
    
👉 [pandas.DataFrame](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html)


### Create a DataFrame From A 2D-ndarray

```python
df = pd.DataFrame(
    [[4, 7, 10],
     [5, 8, 11],
     [6, 9, 12]],
    index=['row_1', 'row_2', 'row_3'],
    columns=["col_a", "col_b", "col_c"]
)
df
```

### Create DataFrame From Series

```python
apples  = pd.Series(data=[1, 2, 3], index=["id1", "id2", "id3"])
oranges = pd.Series(data=[4, 5, 6], index=["id1", "id2", "id3"])
```

## Grouping

Very close to GROUP BY in **<mark>SQL</mark>**. It's a 3-step process:

   * **Split**: a DataFrame is split into groups depending on keys
   * **Apply**: an **aggregative** function (<mark>sum</mark>, <mark>mean</mark>, etc.) is applied on each group
   * **Combine**: result from the previous operations are merged (i.e. reduced) into one new DataFrame


   👉 [Pandas Doc: Comparaison with SQL](https://pandas.pydata.org/docs/getting_started/comparison/comparison_with_sql.html)
   
   
   

🤔 **Which region of the world is the most populated?**

```python
regions = data.groupby("Region")
```

```python
regions[["Population", "Area (sq. mi.)"]].sum().sort_values(by="Population", ascending=False)
```
