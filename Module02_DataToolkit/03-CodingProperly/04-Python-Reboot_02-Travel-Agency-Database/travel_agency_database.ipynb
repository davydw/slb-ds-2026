### Create a function

Create a function `get_key_value_from_city` that takes two arguments `city` and `key` and outputs the corresponding value for the city.

```python
# create the function with the selected arguments
def get_key_value_from_city(city, key):

  # to be completed
  # set default output
    output = f'{key} not defined'

  # to be completed
  # set the output if the required key exists in the selected city
    if key in city.keys():
        output =  city[key]

    return output
# to be completed
# test the function with the given example

# print(get_key_value_from_city(paris_city,"total_population"))
# print(get_key_value_from_city(paris_city,"country"))

print(f'Total population = {get_key_value_from_city(paris_city,"total_population")}')
print(f'Country = {get_key_value_from_city(paris_city,"country")}')
```

### Updating values


Sometimes, city information can be incorrect or requires updates.

Create a `modify_city_key_value` function that replaces an existing key's value with a new one. \
It takes three arguments: a `city_name`, a `key` and a new `value` to assign to the key.

The function will generate a text message indicating one of the following scenarios:

- If the city doesn't exist or is not covered: "*city* doesn't exist or is not covered."
- If the provided key doesn't exist: "*key* doesn't exist."
- If the key has been successfully updated for the city: "*key* has been updated for *city*."

```python
# create the function with the selected arguments

def modify_city_key_value(city_name, key, value):

  # to be completed
  # set default output

    output = f"{city_name} doesn't exist or is not covered."

  # to be completed
  # start the loop over cities

    for city in cities:

    # to be completed
    # set the different outputs depending on whether the key exists

        if city['name'] == city_name:

            if key in city.keys():
                city[key]= value
                output = f"{key} has been updated for {city_name}."

            else:
                output = f"{key} doesn't exist."

    return output
# to be completed
# test the function with examples

# print(modify_city_key_value("Paris","total_population",300000))
# print(modify_city_key_value("New York","total_population",30000))
# print(modify_city_key_value("Chengdu","GDP",300000))


print(f'Paris, Total population = {modify_city_key_value("Paris","total_population",300000)}')
print(f'New York, Total population = {modify_city_key_value("New York","total_population",30000)}')
print(f'Chengdu, GDP = {modify_city_key_value("Chengdu","GDP",300000)}')
```

```python
# create the function with the selected arguments

def get_cities_nearby(customer_location,radius):

  # to be completed
  # start a list
    cities_list=[]

  # to be completed
  # start the loop over cities

    for city in cities:

    # to be completed
    # define the condition on customer_location

        if distance.distance(customer_location,city['location']).km <= radius :

        # to be completed
        # append the list of close cities

            cities_list.append(city["name"])

    return cities_list
# to be completed
# test the function with the example

get_cities_nearby([32.4355, 105.8436],500)
```
