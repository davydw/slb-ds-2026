### Task 5 - Critical Analysis

Create an `"Overrated/Underrated"` column

We have two scores in our DataFrame - `"Rating"` and `"Metascore"`. `"Metascore"` has reviews from the site Metacritic and represents the professional critics' opinion of a movie. `"Rating"` on the other hand, is how IMDb users found the movie. 

Create a new column in your DataFrame which should have the `string` "Overrated" if the IMDB `"Rating"` (x10 so you can compare) is lower than the `Metascore` rating column and "Underrated" if the audience IMDB `"Rating"` is higher than the critic review. 

Hint: You can look into either [iterrows](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iterrows.html) or using the [apply](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html) function along each row. Check the documentation! Or you could try using boolean indexing to do this in two lines for extra points!

When you're done, save the `"Overrated/Underrated"` series you created to a variable called `over_under`

_**Bonus Challenge: How many movies were overrated? How many were underrated?**_

```python
def overrated_func(row):
    if row["Metascore"] > row["Rating"] * 10:
        return "Overrated"
    return "Underrated"
```

```python
df["Overrated/Underrated"] = df.apply(overrated_func, axis = 1)
```

```python
df.head()
```

```python
df["Overrated/Underrated"] = "Underrated"
df.loc[df["Metascore"] > df["Rating"] * 10, "Overrated/Underrated"] = "Overrated"
```

```python
over_under = df["Overrated/Underrated"]
over_under
```

```python
# BONUS - how many of each are there?
over_under.value_counts()
```
