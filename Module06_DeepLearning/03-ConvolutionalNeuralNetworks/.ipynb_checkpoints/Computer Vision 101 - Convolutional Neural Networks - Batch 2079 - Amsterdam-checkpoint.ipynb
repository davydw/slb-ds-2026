## 📚 Libraries

```python
# Data manipulation
import pandas as pd
import numpy as np

# Data Viz
import matplotlib.pyplot as plt
import seaborn as sns

# Sklearn
from sklearn.model_selection import train_test_split

# Dataset
from tensorflow.keras.datasets import fashion_mnist

# Keras
from tensorflow.keras import models, layers, optimizers, regularizers, callbacks, utils
from tensorflow.keras.applications import VGG16
# Classification metrics
from sklearn.metrics import confusion_matrix, classification_report
```

## 👗 The Fashion Dataset from Zalando

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
np.unique(y_train, return_counts=True)
```

```python
x_train_expanded = np.expand_dims(x_train, axis=-1)
x_test_expanded = np.expand_dims(x_test, axis=-1)
```

```python
y_train_encoded = utils.to_categorical(y_train, num_classes = 10)
y_test_encoded = utils.to_categorical(y_test, num_classes = 10)
```

## 🧠 My First CNN

### 🛠️ Architecture and compilation

```python
def my_first_cnn_model():

    #####################
    # 1 - ARCHITECTURE  #  
    #####################
    model = models.Sequential()

    # STEP 0 - PREREQUISITES - INPUT + RESCALING
    model.add(layers.Input(shape=(28,28,1)))
    model.add(layers.Rescaling(1/255.))

    # STEP 1 - CONV LAYERS - MOST IMPORTANT
    
    ##### Conv. Layer. no1     
    model.add(layers.Conv2D(6, kernel_size = (3,3), activation="relu"))
    ##### Conv. Layer. no2     
    model.add(layers.Conv2D(4, kernel_size = (3,3), activation="relu"))
    model.add(layers.MaxPool2D(pool_size = (2,2)))
    ##### Flattening before feeding the Dense layers
    model.add(layers.Flatten())

    # STEP 2 - HIDDEN LAYERS - OPTIONAL BUT RECOMMENDED
    model.add(layers.Dense(16, activation="relu"))
    model.add(layers.Dense(8, activation="relu"))              
                               
    # STEP 3 - PREDICTIVE LAYER
    model.add(layers.Dense(10, activation = "softmax"))

    #####################
    # 2 - COMPILATION  #  
    ##################### 
    model.compile(loss = "categorical_crossentropy",
                  optimizer = "adam",
                  metrics = ["accuracy"]
                 )
    return model
```

```python
model =  my_first_cnn_model()
model.summary()
```

### 🥋 Train

```python
es = callbacks.EarlyStopping(patience = 5, restore_best_weights=True)

history = model.fit(x_train, y_train_encoded,
                    validation_split = 0.20,
                    batch_size = 64, 
                    epochs = 100,
                    verbose = 1,
                    callbacks = [es]
                   )
```

### 🔮 Predictions and confusion matrix

```python
y_pred_probas = model.predict(x_test)
y_pred_probas.shape
```

```python
y_pred = np.argmax(y_pred_probas, axis = 1)
```

```python
sns.heatmap(np.round(confusion_matrix(y_test, y_pred),2), annot = True)
```

## 🧙🏼‍♀️ Using a pretrained model

```python
def my_pretrained_cnn_model():

    #####################
    # 1 - ARCHITECTURE  #  
    #####################
    model = models.Sequential([] )

    # STEP 0 - PREREQUISITES - INPUT + RESCALING
    model.add(layers.Input(shape=(224,224,3)))
    model.add(layers.Rescaling(1/255.))

    # STEP 1 - CONV LAYERS DOWNLOADED FROM TENSORFLOW
    vgg = VGG16(weights='imagenet',
           include_top = False,
           input_shape=(224,224,3)
           )

    vgg.trainable = False

    model.add(vgg)

    ##### Flattening before feeding the Dense layers
    model.add(layers.Flatten())

    # STEP 2 - HIDDEN LAYERS - OPTIONAL BUT RECOMMENDED
    model.add(layers.Dense(16, activation="relu"))
    model.add(layers.Dense(8, activation="relu"))              
                               
    # STEP 3 - PREDICTIVE LAYER
    model.add(layers.Dense(5, activation = "softmax"))

    #####################
    # 2 - COMPILATION  #  
    ##################### 
    model.compile(loss = "categorical_crossentropy",
                  optimizer = "adam",
                  metrics = ["accuracy"]
                 )
    return model
```

```python
model2 =  my_pretrained_cnn_model()
model2.summary()
```
