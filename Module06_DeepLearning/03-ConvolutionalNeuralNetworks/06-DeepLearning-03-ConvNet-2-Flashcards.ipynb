<details>
    <summary>What is the purpose of a Convolutional Neural Network? Can you give some examples of CNN's applied in real-life scenarios?</summary>

Convolutional Neural Network, or CNN for short, refers to a specific type of Neural Network architecture designed to study the patterns in images.

It is widely used in Object detection, Facial recognition, Medical Image Segmentation, Image Classification, etc...
</details>

<details>
    <summary>My image dataset X_train has a shape (1200, 255, 255, 3). What do these dimensions correspond to?</summary>

- 1200 images
- of size 255 (height) * 255 (width)
- with 3 color channels corresponding to the respective intensities of the color values used to create each pixel

For example, the most known system of channels is: RGB (Red Green Blue).
</details>

<details>
    <summary>When dealing with a CNN to classify colored images, the output shape of my CNN hidden layer n°4 is (None, 32, 32, 128). What do these dimensions correspond to?</summary>

- None: is a placeholder for the number of samples in each batch that will be passed through the model.
- 32 * 32 are the spatial dimensions of the images coming out of this layer
- 128 is the number of channels in these images. In the entry layer 3 channels means 3 colors, but in the intermediary layers we can no longer think of them as "colors" but rather "features".
</details>

<details>
    <summary>In any Machine Learning/Deep Learning algorithm, it is recommended, if not mandatory, to standardize/normalize the dataset. How do we do this with images?</summary>

CNNs are usually trained on RGB-coded images (0-255). We can divide the entire dataset by 255 to have all the pixels' intensities lie between 0 and 1.
</details>

Data Augmentation is the process of taking images that you already possess and applying some transformations such as mirror, crop, rotate, etc... to create new versions of the pre-existing images.

<details>
    <summary>If an image is just an array, can we just flatten it into a 1D-vector and use a Dense Neural Network? Explain your answer.</summary>

There is nothing or no one preventing you from doing so, but... :
- It would end up with too many parameters
- Most importantly, an image conveys the same global information even if you rotate it, crop it a bit, mirror it...
- We call this invariance by translation.
- A Convolutional Neural Network can capture this invariance by translation, while a Dense Neural Network cannot.
</details>

<details>
    <summary>In CNNs, what is the difference between kernels and filters?</summary>

- You can think about kernels as magic glasses scanning an individual channel (2D-array) of an image, to extract spatial information/patterns from it.
- A filter contains as many kernels as there are channels in the image to process (ex: 3 channels in colored image, more in intermediary layers)

</details>

<details>
    <summary>What is the approximate shape of a image of size (128, 128) after passing through a Conv2D layer with <b>strides</b> of (2,2)</summary>

Half as long in each dimension: (64, 64). For the given convolutional layer, strides indicate the size of the steps the kernels take while scanning the images.
</details>

<details>
    <summary>What does the "padding" argument in a Conv2D layer correspond to?</summary>

- It is by default deactivated, which means that the pixels on the border of a picture are disadvantaged because of the nature of convolutional operations.
- For images of landscapes, mountains, beaches, you might not need it, but for microscopic medical images it can have a gigantic impact on capturing the patterns along the edges.
- This has a dramatic impact for example when analyzing images of cells being at risk of cancer.
</details>

<details>
    <summary>What is the concept of applying a pooling operation to the output of a convolutional layer?</summary>

- It helps by reducing the size of the output.
- You can consider for example the maximum intensity (MaxPooling) or the average intensity (AveragePooling) of pixels in different areas of a picture.
</details>

<details>
    <summary>What is Transfer Learning?</summary>

- Transfer learning is a Deep Learning technique where a pre-trained model is re-used as a building block/starting point for a new model. When the new model's additional layers are trained on a different task the original model's weights may or may not be updated.
- It's pretty efficient to use start-of-the-art pre-trained convolutional layers from big companies and universities and to train only the dense layers at the end by yourself.
</details>

<details>
    <summary>What is an Autoencoder? What are your features `X` and targets `y` when using an autoencoder? What are they used for?</summary>

1. It is a specific kind of Neural Network (commonly convolutional) with two parts:
  - `encoder`: in this first step, the goal is to compress the images. Eg. From size 224 x 224 x 3 (150k numbers) into a small vector of size 8 (the size is called the LATENT DIMENSION).
  - `decoder`: in this second step, the goal is to decode the compressed data and hopefully retrieve the original images.

2. As we try to minimize the loss between our outputs and our inputs, we fit like this: `autoencoder.fit(X_train, X_train)`

3. They can be used for:
  - Compression (trying find the lowest latent dimensions that is still able to capture most of the information).
  - Using only the encoder: e.g. represent words into fixed size vectors given by the latent space.
  - Using only the decoder: synthesize new outputs (like deep fakes) by randomly sampling from the latent space and using only the decoder! (Generative Models).
</details>
