<details>
    <summary>Why is it necessary to Compile a Neural Network? What are the three components of <b>.compile(...)</b> and what are their respective roles?</summary>

Compiling is required to inform Keras how to optimize the different parameters defined in the layers:

* **loss**: which loss function are we trying to minimize DURING the training phase
* **metrics** : which metric(s) are we going to use to evaluate the model AFTER the training procedure
* **optimizer** : which advanced numerical method will be used to minimize the loss function

</details>

<details>
    <summary>According to which objectives are the parameters of a Neural Network updated?
Name a few methods for updating these weights</summary>

* Parameters of a Neural Network are updated (optimized) with respect to the Loss Function, which we aim to minimize.
* In many cases, the Loss Function is extremely hard to minimize. This is why we use iterative methods, closely related to stochastic gradient descents (SGD)
* More complex optimizers include `adam` and `rmsprop`.

</details>

<details>
    <summary>What is Regularization in Deep Learning?</summary>

Although powerful, remember that Dense Neural Networks are nothing other than a bunch of “activated linear regressions”.
Hence, in a similar fashion to Machine Learning, we can apply L1/L2 penalties to these parameters.
</details>

<details>
    <summary>What is the learning rate in neural network?</summary>


During every backward propagation, the weights are adjusted **proportionally** to the derivative of the loss function. This proportion is quantified through the learning rate  $\alpha$ (“alpha”).
</details>

<details>
    <summary>Why can’t we use the test set to decide when to stop the training? Which set should we use instead?</summary>

* Never use the test set to stop the training:
    * First of all, you are not supposed to see the test set during the training phase.
    * Second, you might not even have access to it.
    * Last but not least, you are leaking your data!
* It is a good practice to split the data into three sets: Training, Validation, and Testing. The Stopping Criterion should be set on the validation set.

</details>

<details>
    <summary>My train loss does not improve by an inch, even after many epochs! What should I do with my learning rate, which is set a default value now?

</summary>

* Try to increase it until your model starts to “learn” something!
* A more advanced strategy is to use a learning rate that changes over time. eg. Exponential Schedulers progressively decrease the learning rate so that the model learns quickly initially but takes smaller steps later in the training so it doesn’t potentially “jump over” the minimum of the loss function

</details>

<details>
    <summary>Would you rather apply Dropout/Regularization to the first layers of the Neural Network or the last?</summary>

In general, it is not a good idea to apply regularizers or dropout to the first layers. Think about a child, you want to let him/her develop his/her creativity during the beginning of their life before giving them a good framework and rules. The same applies to Neural Networks: you want it to be able to learn quite fast initially before applying Regularizers and Dropout during the last layers.

</details>

<details>
    <summary>What is the role of a Dropout layer?</summary>

* The role of a Dropout layer is to cancel out temporarily the role of some neurons in a forward/backward propagation
* The word “temporarily” is important here: by doing this, the Neural Network doesn’t update some parameters during one forward/backward propagation and hence prevents them from overspecializing

</details>

<details>
    <summary>Cite the three main techniques to prevent overfitting in Dense Neural Networks</summary>

1. Early Stopping
2. Adding Regularization to layers
3. Dropout layers

</details>

<details>
    <summary>Can you explain what forward propagation and backward propagation are?</summary>

* **Forward propagation**: The network takes a batch of data points as inputs, which go through the different layers, to output some PREDICTIONS based on the current values of the weights and biases.
    * Loss computation: The model then computes the LOSS between the real values in this batch versus the previously predicted values.
** **Backward propagation**: The weights and biases are ADJUSTED so as to reduce the loss computed above, starting with the parameters of the last layer, and going all the way back to the deepest layer.

</details>
