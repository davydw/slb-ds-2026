<details>
    <summary>What activation function would you use in the case of a multi-class classification?</summary>

For an n-class classification, you need n output neurons, each outputting a probability of belonging to the class 
 with ($k$ for $ 1\leq k \leq n$).
To do so, you can use a softmax activation layer.    
</details>

<details>
    <summary>What’s the difference between Machine Learning and Deep Learning? On what types of data do Neural Networks perform particularly well?</summary>

Machine learning means computers learning from data using algorithms to perform a task without being explicitly programmed. Deep learning is a subset of Machine Learning, and uses a complex structure of algorithms modeled on the human brain. This enables the processing of unstructured data such as documents, images, and text.    
</details>

<details>
    <summary>What is an epoch in Neural Network training?</summary>
    
In terms of Artificial Neural Networks, an epoch refers to one cycle through the full training dataset.    
</details>

<details>
    <summary>If you are performing a two-class classification, what type of activation function does your Neural Network need for the output neuron?</summary>

In the case of a two-class classification, the output neuron must output a probability (of belonging to either class). Sigmoid activation functions offer such an output.   
</details>

<details>
    <summary>Why do we use activation functions in Neural Networks ?</summary>

Without any activation function, the predictive neuron would just be a linear combination of linear combinations, that is… still a linear combination!    
</details>

<details>
    <summary>What are the drawbacks of Neural Networks?</summary>

* Intractable mathematical solutions (The hypotheses and cost functions are too complex to be solved analytically)
* 👉 Reliance on heuristics, and trial & error. Still in early stages of fine-tuning strategies!
* Computationally expensive to train (several orders of magnitude more parameters to find than conventional ML)
* Models don’t converge without a big dataset
Real lack of interpretability:
    * in Machine Learning, especially in Regression Models, you can interpret the role of some coefficients…
    * in Deep Learning, it’s difficult to say “well this neuron is dedicated to…”    
</details>

<details>
    <summary>What is a “neuron”, the key element of an Artificial Neural Network?</summary>
    
* A neuron is a mathematical function: a linear combination of different features + a bias (think linear regression), wrapped in an activation function.
* It is modeled after the functioning of a biological neuron.    
</details>

<details>
    <summary>In Neural Networks, what are activation functions?</summary>

Activation functions are mathematical equations that determine the output of a neural network. The function is attached to each neuron in the network, and determines whether it should be activated or not, based on whether each neuron’s input is relevant for the model’s prediction.    
</details>
