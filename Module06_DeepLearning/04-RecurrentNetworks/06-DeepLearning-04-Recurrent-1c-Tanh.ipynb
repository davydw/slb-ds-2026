```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10,10.1,0.1)
```

```python
y = np.tanh(x)
```

```python
y_first_derivative = 1 - (np.tanh(x))**2
```

```python
y_second_derivative = - 2 * np.tanh(x) / np.cosh(x)
```

```python
plt.figure(figsize=(8,8))
plt.plot(x,y, c="green", label = "tanh")
plt.plot(x,y_first_derivative, c="blue", label = "first derivative", linestyle="dashdot")
plt.plot(x,y_second_derivative, c="red", label = "second derivative", linestyle="dotted")
plt.grid(visible = True)
plt.legend(loc="best")
plt.show()
```
