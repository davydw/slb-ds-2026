<details>
    <summary>What type of data are Recurrent Neural Networks used for?</summary>

Recurrent Neural Networks are designed to predict using sequential (temporal) input data.
</details>

<details>
    <summary>One-to-One and One-to-Many are examples of possible Neural Network tasks. Name a few others.</summary>

* Many-to-one
* Many-to-many
</details>

<details>
    <summary>Name two different architectures of Recurrent layers.</summary>

* LSTM: Long Short Term Memory
* GRU: Gated Recurrent Units
</details>

<details>
    <summary>Why was LSTM architecture introduced?</summary>

LSTM architecture was introduced to prevent the "vanishing gradient"
</details>

<details>
    <summary>Why was GRU architecture introduced?</summary>

GRU architecture was introduced to decrease the number of parameters currently used in LSTMs. It's aim is faster training times, with potentially less training data.
</details>
