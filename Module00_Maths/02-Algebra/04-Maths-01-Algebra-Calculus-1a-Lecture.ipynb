### Functions

A function is a relation between sets that associates every element of the first set to **exactly one element** of the second set.

Example:

$$
f\colon \mathbb{R} \to \mathbb{R}, f(x) = x^2 + 1
$$

In Python, we would implement this function with **one parameter** (`int`), returning a value (`int`):

```python
def square_plus_one(x):
    return x * x + 1
```

In the rest of the course, we'll assimilate the $f(x)$ notation with $y$

### Multi-variate functions

A function can be defined by more than two variables, like:

$$
f\colon \mathbb{R^2} \to \mathbb{R}, f(x_0, x_1) = 2 x_0 + 3 x_1 + 7
$$

which will translate in the Python world by:

```python
def multi_variate(x0, x1):
    y = 2 * x0 + 3 * x1 + 7
    return y
```

### Inverse Function

It is a function that "reverses" another function: if the function $f$ applied to an input $x$ gives a result of $y$, then applying its inverse function $g$ to $y$ gives the result $x$, and vice versa.

$$
f(x) = y \iff g(y) = x
$$

⚠️ Not all functions have an inverse. If it exists, $g$ is often denoted as $f^{-1}$ and both $f$ and $f^{-1}$ are [**bijections**](https://en.wikipedia.org/wiki/Bijection).

Example:
    
$$
f(x) = \frac{3}{2}x + 3
$$

$$
g(y) = \frac{2}{3}y - 2
$$

$$
g(f(x)) = \frac{2}{3}(\frac{3}{2}x + 3) - 2 = x
$$

## 2) System of linear equations

A system of linear equations (or **linear system**) is a collection of one or more linear equations involving the same set of variables.

$$
\begin{cases}
\begin{alignedat}{7}3x&&\;+\;&&2y&&\;-\;&&z&&\;=\;&&1&\\2x&&\;-\;&&2y&&\;+\;&&4z&&\;=\;&&-2&\\-x&&\;+\;&&{\tfrac {1}{2}}y&&\;-\;&&z&&\;=\;&&0&\end{alignedat}
\end{cases}
$$

A **solution** to a linear system is an assignment of values to the variables such that all the equations are simultaneously satisfied. For instance:

$$
{\begin{alignedat}{2}x&\,=\,&1\\y&\,=\,&-2\\z&\,=\,&-2\end{alignedat}}
$$

The word **system** indicates that the equations are to be considered collectively, rather than individually. 

### Trivial example (one unknown)

$$
2x = 4
$$

### Simple non-trivial example (two unknowns)

🤔 Problem:

- A father is 22 years older than his son.
- In 10 years, the father will be twice as old as his son.

How old are they?

✏️ Let $y$ be the father's age, $x$ the son's age. Let's solve this on the whiteboard!

We can translate the problem into an equation with the following system:

$$
\begin{cases}
\begin{alignedat}{5}
&&y\;-\;&&x&&\;=\;&&22&\\
&&y\;-\;&&2x&&\;=\;&&10&
\end{alignedat}
\end{cases}
$$

And can solve this **graphically**!

```python
x = np.array([6, 18])
plt.plot(x, x + 22)
plt.plot(x, 2 * x + 10)
plt.xlabel('son')
plt.ylabel('father')
plt.show()
```

### Euclidian Distance 
$\lVert \vec{AB}\rVert = L^{2}(\vec{AB})$ for "$L^{2}$ distance"

✏️ Let's go to the whiteboard and draw two points $A = (2, 4)$ and $B = (3, 2)$.

Using the [Pythagorean theorem](https://en.wikipedia.org/wiki/Pythagorean_theorem), compute the distance between $A$ and $B$.

$$
d = \sqrt {\left( {x_A - x_B } \right)^2 + \left( {y_A - y_B } \right)^2 }
$$

The distance formula can be extended to **more dimensions**.

Let $X$ and $Y$ be two points of a normed vector space of dimension $n$

$$
A=(a_{1},a_{2},\ldots ,a_{n}){\text{ and }}B=(b_{1},b_{2},\ldots ,b_{n})\in {\mathbb  {R}}^{n}
$$

The Euclidian Distance between those two points is defined by:

$$
{\displaystyle L^{2}(\vec{AB})=||\vec{AB}||=\sqrt{\sum _{i=1}^{n}|b_{i}-a_{i}|^{2}}}
$$

Sometimes we need the $L^{1}$ distance which is defined as:

$$
L^{1}(\vec{AB})=|\vec{AB}|=\sum _{i=1}^{n}|b_{i}-a_{i}|
$$

### Matrix Multiplication 🔥

Let's have a look at the _formal definition_ 🤯

Multiplication of two matrices is defined iif the **number of columns of the left matrix** is the same as the **number of rows of the right matrix**.

If $A$ is an $m \times n$ matrix and $B$ is an $n \times p$ matrix, then the result of their product $C = AB$ is the $m \times p$ matrix whose entries are given by the **dot product** of the corresponding row of $A$ and the corresponding column of $B$:

$$
   {\displaystyle [\mathbf {AB} ]_{i,j}=a_{i,1}b_{1,j}+a_{i,2}b_{2,j}+\cdots +a_{i,n}b_{n,j}=\sum _{r=1}^{n}a_{i,r}b_{r,j},}
$$

### Solving a linear system

Remember the father/son ages problem? Let $x_1$ be the father's age, and $x_2$ the son's age:

$$
\begin{cases}
\begin{alignedat}{5}
&&x_1\;-\;&&x_2&&\;=\;&&22&\\
&&x_1\;-\;&&2x_2&&\;=\;&&10&
\end{alignedat}
\end{cases}
$$

✏️ Let's go to the whiteboard and re-write the equation as a matrix equation, $AX = B$

$$
\begin{bmatrix}
1 & -1 \\
1 & -2 \\
\end{bmatrix}
\begin{bmatrix}
x_1 \\
x_2 \\
\end{bmatrix}
=
\begin{bmatrix}
22 \\
10
\end{bmatrix}
$$

$$
AX=B
$$

We can multiply both sides of the equation with $A^{-1}$:

$$
A^{-1}AX=A^{-1}B
$$

By definition, $A^{-1}A = I$, so the equation get simplified with:

$$
X=A^{-1}B
$$

Let's try to do it with NumPy:

```python
A = np.array([[1, -1], [1, -2]])
A_inv = np.linalg.inv(A)
A_inv
```

```python
B = np.array([[22], [10]])
B
```

```python
np.matmul(A_inv, B)
```

You can do it directly using the [`numpy.linalg.solve`](https://docs.scipy.org/doc/numpy/reference/generated/numpy.linalg.solve.html) method:

$$
AX=B
$$

```python
A = np.array([[1, -1], [1, -2]])
B = np.array([[22], [10]])
X = np.linalg.solve(A, B)
X
```

### How did Numpy compute $A^{-1}$ ?

Let's write $A^{-1}A = I$, where dimension of matrices are $n = 2$:

$$
\begin{bmatrix}
a_{1,1} & a_{1,2} \\
a_{2,1} & a_{2,2}
\end{bmatrix}
\begin{bmatrix}
1 & -1 \\
1 & -2
\end{bmatrix}
=
\begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}
$$

We can use the [**Gaussian elimination**](https://en.wikipedia.org/wiki/Gaussian_elimination#Finding_the_inverse_of_a_matrix) algorithm which has a $O(n^3)$ **complexity**. 

This complexity can still be [improved](https://en.wikipedia.org/wiki/Computational_complexity_of_mathematical_operations#Matrix_algebra) but we won't reach a complexity that is better than $O(n^{2.3})$ 😱
