<details>
  <summary>Can you explain the concept of inverse function?</summary>

Imagine that you are considering a function $f$ that maps all the elements $x$ of a set $X$ of inputs to the elements $y$ of a set $Y$ of outputs.

The inverse function $f^{-1}$ is the function that maps all the elements $y$ of the set $Y$ to the elements $x$ of the set $X$.

For example, let's consider $ X = [0, +\infty ]$, the set of all positive input numbers and $ Y = [0, +\infty ]$, the set of all positive output numbers.

Let's define the function $f$ such as $y = f(x) = x^2$.
Then, we can define its inverse function $f^{-1}$ by $x = f^{-1}(y) = \sqrt{y}$

Please be aware that many functions don't have an inverse function and that for some functions, you need to be very careful with the domain of definition $X$ and the codomain of definition $Y$!
</details>

You should convert the problem into a system of linear equations and re-arrange it to make some functions appear:

<details>
  <summary>What are the three possible scenarios when it comes to solving $X$ in system of linear equations $AX = B$?</summary>

1. There can be no solution. For instance, if $A$ is the null matrix and $B$ a column vector full of ones, there is no $X$ that solves $AX = B$
2. There can be one solution $X=A^{-1}B$ if $A$ is invertible
3. There can be an infinite number of solutions. For example, any vector $X$ would solve $AX = B$ if $A$ is a null matrix and $B$ is a null vector!
</details>

<details>
  <summary>What is the inverse of a matrix? Does it always exist?</summary>

Let's start with the real numbers. All of them have an `inverse` except for 0. For example, the inverse of 3 is $\frac{1}{3}$ because $ 3 \times \frac{1}{3} = \frac{1}{3} \times 3 = 1$.

How does it work with matrix? We say that a matrix $A$ is invertible if there exists a matrix $B$ such that $ AB = BA = I$ where $I$ is the identity.

The inverse matrix is called $B = A^{-1}$ but it doesn't always exist. A first requirement is that the matrix $A$ be square
</details>

**The derivative of a function represents its rate of change with respect to its argument.**
      For example, let's say that you buy a premium smartphone for 500 EUR and you subscribe to a mobile plan for 20 EUR a month.
      The cost related to your phone can be written as $ f(m) = 500 + 20 \times m $ where $m$ is the number of months.
