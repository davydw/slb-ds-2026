# 02-DataToolkit_04-StatsProbas-Flashcards

<details>
  <summary>What is statistical inference?</summary>

Statistical inference is drawing statistical conclusions from specific data, using probability.
</details>

<details>
  <summary>What is a histogram?</summary>

A histogram is a representation of the distribution of numerical data.

It is an estimate of the probability distribution of a continuous variable.
</details>

<details>
  <summary>What is the median of a data sample?</summary>

The median is the value separating the upper half from the lower half of a data sample.
</details>

<details>
  <summary>What is the equation that defines the mean of a sample?</summary>

The mean of a **sample** of the population $x_1, x_2, ..., x_n$ ($n < N$) is defined by:

$${\bar {x}} = {\frac {1}{n}}\sum _{i=1}^{n}x_{i}={\frac {x_{1}+x_{2}+\cdots +x_{n}}{n}}$$
</details>

<details>
  <summary>What data transformation is required before plotting histograms?</summary>

The data needs to be binned. The bins (intervals) must be adjacent, and are often of equal size.
The height of each bin will be the count (or frequency) of data points falling in that bin.

[Source](https://en.wikipedia.org/wiki/Histogram)
</details>

<details>
  <summary>Let’s consider this random experiment: you draw two cards, one at a time, without replacement, in a deck of 52 cards.

You draw the first card: it’s a King (event 𝐴). You draw a second card: it’s a King too (event 𝐵)

* What’s the conditional probabilty of 𝐵 knowing 𝐴 happened?
* What’s the probability of 𝐴 ∩ 𝐵 of having drawn two kings?</summary>

$$P(A) = \frac{4}{52} \approx 0.077$$

$$P(B \mid A) = \frac{3}{51} \approx 0.059$$

$$P(A \cap B) = P(A) \cdot P(B\mid A) = \frac{1}{221} \approx 0.005$$
</details>

<details>
  <summary>How would you define a random experiment?</summary>

A random experiment is a process by which we observe something uncertain.
After the experiment, the result of the random experiment is known: it is the outcome.

The set of all possible outcomes is called the sample space 𝑆.
</details>

<details>
  <summary>What is statistical dispersion?</summary>

Dispersion (also called variability, scatter, or spread) is the extent to which a distribution is stretched or squeezed.
</details>

<details>
  <summary>What is the equation for Bayes’ Theorem?</summary>

$$P(A\mid B)={\frac {P(B\mid A) \cdot P(A)}{P(B)}}$$ 

for any two events $A$ and $B$, where $P(B) \neq 0$:
</details>

<details>
  <summary>If we consider an entire soccer match as a random experiment, what would be potential random variables?</summary>

Goals, shots, shots on target and possesion are all examples of random variables.
</details>

<details>
  <summary>What is the mode? Is it always unique?</summary>

The mode is the most frequent value of a set. A dataset can be multimodal.
</details>

<details>
  <summary>In statistics, what is the variance and its corresponding equation?</summary>

The variance measures how far a set of random numbers are spread out from their average value.

The **variance** of a population of $N$ elements is:
$$ {\displaystyle \sigma^2={{\frac {1}{N}}\sum _{i=1}^{N}(x_{i}-{\mu})^{2}}} $$
</details>

<details>
  <summary>What is the expected value of a random variable?</summary>

$X$ being a **discrete** random value with a finite number of finite outcomes $x_1$, $x_2$, ..., $x_k$ occuring with probabilities $p_1$,$p_2$, ..., $p_k$ respectively. 

$${\displaystyle \operatorname {E} [X]=\sum _{i=1}^{k }x_{i}\,p(x_i).}$$
</details>

<details>
  <summary>What is the difference between mean and median in respect to outliers?</summary>

The median is robust against outliers, which is not the case for the mean.
</details>

<details>
  <summary>What is the difference between a histogram and a bar chart?</summary>

A histogram is used to represent continuous univariate data and a bar chart represents categorical and numerical data.
</details>

<details>
  <summary>What is the Interquartile range and what can it be used for?</summary>

It is the difference between upper and lower quartiles.

$$IQR= Q_3-Q_1$$

It can be used to identify outliers in the data set.
</details>
