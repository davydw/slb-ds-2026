<details>
  <summary>How do you call a method on a string?</summary>

With a `.` and the name of the method, after the given string.
For instance, calling the `.upper()` method on the "Hello World" string would give:
```python
In [1]: "Hello World".upper()
Out[1]: "HELLO WORLD"
```
</details>

<details>
  <summary>How do you convert a str into an int?</summary>

You can use the `int()` function on it:
```python
In [1]: int('42')
Out[1]: 42

In [2]: type(int(42))
Out[2]: int
```
</details>

<details>
  <summary>How do you Read a value in a dict?</summary>

```python
In [1]: instruments = {'john': 'guitar', 'paul': 'bass'}

In [2]: instruments['paul']
Out[2]: 'bass'

In [3]: instruments['ringo']
# => KeyError: 'ringo'

In [4]: instruments.get('ringo')
Out[4]: None
```
</details>

<details>
  <summary>How do you Update a value in a dict?</summary>

```python
In [1]: instruments = {'john': 'guitar', 'paul': 'bass', 'george': 'guitar'}

In [2]: instruments['george'] = 'sitar'

In [3]: instruments
Out[3]: {'john': 'guitar', 'paul': 'bass', 'george': 'sitar'}
```
</details>

<details>
  <summary>What is string interpolation and how do you use it?

</summary>

String interpolation allows you to insert code into strings. The code will be replaced by its result.
```python
In [1]: f'4 + 3 = {4+3}'
Out[1]: '4 + 3 = 7'
```
or
```python
In [2]: name = "Toto"
In [3]: f'My name is {name}'
Out[3]: 'My name is Toto'
```
</details>

<details>
  <summary>What function can you call on an object to determine its type (<b>str</b>, <b>int</b>, <b>list</b>, ...)?</summary>

You can call the `type()` built-in function on it:
```python
In [1]: type(12)
Out[1]: int

In [2]: type(7.8)
Out[2]: float

In [3]: type("Hello")
Out[3]: str

In [4]: type(["Sponge Bob", 12, 3.14])
Out[4]: list
```

<details>
  <summary>How do you Delete an element from a list?</summary>

```python
In [1]: beatles = ["John", "Paul", "George", "Ringo"]

In [2]: del beatles[0]    # Delete by its index

In [3]: beatles
Out[3]: ['Paul', 'George', 'Ringo']

In [4]: beatles.pop(0)    # Delete by its index and return the value
Out[4]: 'Paul'

In [5]: beatles
Out[5]: ['George', 'Ringo']

In [6]: beatles.remove('George')    # Delete by its value

In [7]: beatles
Out[7]: ['Ringo']
```
</details>

  <summary>How do you loop over a <b>list?</b></summary>

```bash
In [1]: beatles = ["John", "Paul", "George", "Ringo"]

In [2]: for beatle in beatles:
    ...:     print(beatle)
    ...:
John
Paul
George
Ringo
```

<details>
  <summary>How do you loop over a <b>dict?</b></summary>

```python
In [1]: instruments = {'john': 'guitar', 'paul': 'bass'}

In [2]: for name, instrument in instruments.items():
    ...:     print(f'{name.capitalize()} plays the {instrument}')
    ...:
John plays the guitar
Paul plays the bass
```
</details>

<details>
  <summary>How do you Delete a key/value pair from a dict??</summary>

```python
In [1]: instruments = {'john': 'guitar', 'paul': 'bass', 'george': 'guitar'}

In [2]: del instruments['george']

In [3]: instruments
Out[3]: {'john': 'guitar', 'paul': 'bass'}
```
</details>
